### difference between shape of tensors and shape of vectors

In mathematics and computer science, a vector is a one-dimensional array of numbers, while a tensor is a multi-dimensional array of numbers.

The shape of a vector refers to the number of elements in the array and is usually represented as a single number. For example, a vector with 5 elements has a shape of (5,). This shape tuple tells us that the vector has one axis and the length of that axis is 5.

On the other hand, the shape of a tensor refers to the number of elements in each dimension of the multi-dimensional array. For example, a 2D tensor with 3 rows and 4 columns has a shape of (3, 4). This shape tuple tells us that the tensor has two axes, one with a length of 3 and the other with a length of 4.

In summary, while both vectors and tensors can have shapes, the shape of a vector is represented as a single number, whereas the shape of a tensor is represented as a tuple of numbers, with each number indicating the size of a particular dimension of the tensor.

### Mean centering

__Mean centering__ is a useful pre-processing technique in various machine learning
applications. It can improve the performance, convergence, and interpretability of the
model. Some practical examples where mean centering is applied include:


1. __Principal Component Analysis (PCA):__ PCA is a dimensionality reduction technique that
transforms the data into a new coordinate system by identifying the directions (principal
components) with the highest variance. Before applying PCA, it is essential to mean
center the data to ensure that the first principal component represents the direction
with the highest variance in the dataset, rather than being influenced by the location of
the data in the coordinate system.


2. __Linear regression:__ In linear regression, mean centering can help improve the
interpretability of the model coefficients by making them directly comparable. When the
features are mean-centered, the intercept term represents the expected value of the
dependent variable when all independent variables are at their mean values.
Additionally, mean centering can help with multicollinearity issues, especially when
there are interaction terms in the model.


3. __Gradient-based optimization algorithms:__ Some machine learning algorithms, such as
gradient descent, can converge faster when the input features are mean-centered. This
is because mean centering can lead to better conditioning of the optimization problem,
allowing the gradient descent algorithm to take larger, more consistent steps towards
the optimal solution.


4. __Clustering algorithms:__ Mean centering can help improve the performance of clustering
algorithms like k-means by ensuring that the initial cluster centroids are not heavily
influenced by the location of the data in the coordinate system. This can lead to faster
convergence and better clustering results.


5. __Regularization:__ In machine learning models that use regularization techniques, such as
ridge regression or LASSO, mean centering can help ensure that the regularization term
has a consistent effect across all features. By mean centering the features, the model is
less likely to penalize the intercept term, which can lead to better generalization.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate random 2D data with 100 points
data = np.random.rand(100, 2)

# Calculate the mean of the data along each dimension
data_mean = np.mean(data, axis=0)

# Perform mean centering
centered_data = data - data_mean

# Plot the original data
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], label="Original Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Before Mean Centering")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

# Plot the mean-centered data
plt.subplot(1, 2, 2)
plt.scatter(centered_data[:, 0], centered_data[:, 1], label="Mean-Centered Data", color="orange")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("After Mean Centering")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

plt.show()
````

### Rules of Dot Product

#### 1. Commutative

$$A.B = B.A$$


#### 2. Distributive

$$A.(B+C) = A.B + A.C$$

##### commutative example : 

````python
import numpy as np

# Define three vectors
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

# Calculate dot product using numpy.dot() function
dot_product_AB = np.dot(A, B)
dot_product_BA = np.dot(B, A)

print(dot_product_AB)
print(dot_product_BA)

# Check commutative property: A  B = B  A
if dot_product_AB == dot_product_BA:
    print("The dot product is commutative.")
else:
    print("The dot product is not commutative.")


````

##### distributive example

````python
# Define three vectors
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])
C = np.array([7, 8, 9])
````

````python
# Check distributive property: A  (B + C) = A  B + A  C
dot_product_A_BC = np.dot(A, B + C)
dot_product_AB_plus_AC = np.dot(A, B) + np.dot(A, C)

print(dot_product_A_BC)
print(dot_product_AB_plus_AC)


if dot_product_A_BC == dot_product_AB_plus_AC:
    print("The dot product is distributive.")
else:
    print("The dot product is not distributive.")
````

### cosine similarity

- cos of angle between 2 vectors


- how similar 2 vectors are


 
 $$cos\theta = \frac{u.v}{||u||\;||v||}$$
 
 
- 1 means $\theta = 0\;degrees$ which mean both vectors are similar (same directions)
 

- -1 means $\theta = 180\;degrees$ which mean both vectors are very dissimilar


- 0 means $\theta = 90\;degrees$ which means both vectors are orthogonal

````python
import numpy as np

# Define two vectors
A = np.array([1, 2, 3])
B = np.array([-4, -5, -6])
C = np.array([5,5,5])


# Calculate the cosine similarity
cosine_similarity = np.dot(A, B) / (np.linalg.norm(A) * np.linalg.norm(B))

print("Cosine similarity between A and B:", cosine_similarity)

# Calculate the cosine similarity
cosine_similarity = np.dot(A, C) / (np.linalg.norm(A) * np.linalg.norm(C))

print("Cosine similarity between A and C:", cosine_similarity)
````

### Types of Matrices : 

- __Square matrix:__ A square matrix has the same number of rows and columns. For example, a 3x3 matrix is a square matrix because it has three rows and three columns.



- __Identity matrix:__ An identity matrix is a square matrix in which all the diagonal elements are 1, and all other elements are 0. For example, the 3x3 identity matrix is as follows:

        1 0 0
        0 1 0
        0 0 1



- __Diagonal matrix:__ A diagonal matrix is a square matrix in which all the non-diagonal elements are 0. For example, the 3x3 diagonal matrix is as follows:

        2 0 0

- __Column matrix:__ A column matrix is a matrix with only one column. For example, a 3x1 matrix is a column matrix.

### Additive identity

The additive identity is a matrix that, when added to any other matrix, leaves the matrix unchanged.


In other words, it is the matrix equivalent of the number 0 in ordinary arithmetic. The additive identity for matrices is denoted by a matrix with all elements equal to zero, and it is sometimes called the "zero matrix."

$$A+X=A$$

$$X\;\text{is zero Matrix}$$

### Additive Inverse

The additive inverse of a matrix is the matrix that, when added to the original matrix, results in the additive identity (i.e., the zero matrix). 


In other words, it is the matrix equivalent of the negative of a number in ordinary arithmetic. The additive inverse of a matrix is denoted by placing a negative sign in front of the matrix.

$$A+X=0$$

$$X\;is -A$$

## Co-factor :

Cofactor of an element of aij of a determinant is defined by

$$A_{ij} = (-1)^{i+j} M_{ij}$$ 

$$\text{where Mij is the minor of aij}$$

__Every element of a matrix $(a_{ij})$ has a co-factor__

cofactor for each element of the given 3x3 matrix \( A \):


A = \begin{bmatrix}
2 & 4 & 1 \\
3 & 0 & 5 \\
1 & 6 & 2 \\
\end{bmatrix}


**Cofactor Calculation (Step by Step):**

**Element \(2\) (Top Left):**

Step 1: Calculate the minor of element \(2\):

$
\text{Minor of } 2 = \text{det}\begin{pmatrix} 0 & 5 \\ 6 & 2 \end{pmatrix} = (0 \cdot 2) - (5 \cdot 6) = -30
$

Step 2: Apply the cofactor formula:

$
\text{Cofactor of } 2 = (-1)^{1+1} \cdot \text{Minor of } 2 = -30
$

**Element \(4\) (Top Middle):**

$

$

Step 2: Apply the cofactor formula:

$
\text{Cofactor of } 4 = (-1)^{1+2} \cdot \text{Minor of } 4 = -1
$

Continue with the same process for each element. Remember that the cofactor of an element is calculated by multiplying the sign pattern \((-1)^{i+j}\) where \(i\) is the row number and \(j\) is the column number, with the corresponding minor of the element.

### NOTE : Determinant = Sum of the product of elements of any row(or col) with their corresponding cofactors.

**Determinant Calculation Using Cofactors (Step by Step):**

Let's consider the 3x3 matrix \( A \):

$
A = \begin{bmatrix}
2 & 4 & 1 \\
3 & 0 & 5 \\
1 & 6 & 2 \\
\end{bmatrix}
$

**Step 1:** Choose a row or column to start with. For demonstration, let's use the first row.

**Step 2:** Multiply each element in the chosen row (or column) by its corresponding cofactor, and then sum up these products.

For the first row of matrix \( A \):

$
\text{Determinant} = (2 \cdot \text{Cofactor of } 2) + (4 \cdot \text{Cofactor of } 4) + (1 \cdot \text{Cofactor of } 1)
$

Substitute the cofactor values:

$
\text{Determinant} = (2 \cdot -30) + (4 \cdot -1) + (1 \cdot -18) = -60 - 4 - 18 = -82
$

So, the determinant of matrix \( A \) is \(-82\).

This process works for any row or column you choose. You can also use the cofactor expansion along any row or column to calculate the determinant.

## Adjoint

The adjoint of a matrix is a matrix where each element is replaced by its corresponding cofactor, and then the matrix is transposed (rows become columns and columns become rows). The adjugate of matrix A is denoted as adj(A).


To calculate the adjoint of a matrix, follow these steps:

1. Calculate the cofactor of each element in the matrix.
2. Create a new matrix using the cofactors, where each element is in the position of the element for which the cofactor was calculated.
3. Transpose the matrix by swapping rows and columns.

Let's go through an example to calculate the adjoint of a matrix.

**Example: Calculating the Adjoint of a Matrix**

Consider the 3x3 matrix \( A \):


A = \begin{bmatrix}

3 & 0 & 5 \\

\end{bmatrix}

**Step 2:** Create a new matrix using the cofactors.

## Basis vectors  ($\hat i, \hat j, \hat k....$) :

Basis vectors are fundamental vectors that form the building blocks for describing points, directions, or spaces in linear algebra. They serve as reference points or directions in a vector space, allowing us to express other vectors in terms of their linear combinations.

#### in 2-d space:

The 2 vectors $\hat i\;and\;\hat j$ are unit vectors.

- $\hat i$ is for x-axis

any vector can be represented in terms of $\hat i\;and\;\hat j$

For example, in 2D space, any vector **v** can be expressed as **v** = x**i** + y**j**, where **i** and **j** are the basis vectors, and x and y are scalar coefficients. Similarly, in 3D space, any vector **v** can be expressed as **v** = x**i** + y**j** + z**k**.

### This $2\hat i + 3\hat j$ is called Linear Combination

In a vector space, a set of basis vectors has two main properties:

1. **Linear Independence**: The basis vectors are linearly independent, which means that none of the basis vectors can be written as a linear combination of the others.

2. **Spanning**: The basis vectors span the entire vector space, meaning that any vector in the space can be expressed as a unique linear combination of the basis vectors.

## Matrix Multiplication as Composition : 

The composition of transformations in terms of matrices involves applying multiple transformations successively by multiplying their corresponding transformation matrices. This process allows us to combine multiple transformations into a single transformation matrix, making it efficient for calculations and applications.

Here's how the composition of transformations using matrices works:

1. **Individual Transformations**: Each transformation (e.g., translation, rotation, scaling) can be represented by a matrix. These individual transformation matrices capture the effects of the transformations on points or vectors.


2. **Transformation Matrix Multiplication**: To combine transformations, you multiply their corresponding transformation matrices in the order you want to apply them. The resulting matrix represents the composite transformation.


3. **Applying the Composite Transformation**: To transform a point or vector using the composite transformation, you simply multiply it by the composite transformation matrix. This efficiently applies all the desired transformations in the specified order.

For example, if you have a translation matrix **T** and a rotation matrix **R**, and you want to translate a point and then rotate it, you can create a composite transformation matrix **C** by multiplying the matrices: **C = R * T**. When you multiply a point by **C**, it will first be translated by **T** and then rotated by **R**.

__The key advantage of using matrix composition is that you can efficiently perform multiple transformations using a single matrix multiplication__, which is particularly useful in computer graphics, robotics, and other fields where applying sequences of transformations is common.

## Why only square matrix has inverse:

An inverse is possible only for square matrices because it is related to the concept of a matrix
being a bijective linear transformation, which implies both injectivity (one-to-one) and
surjectivity (onto). A square matrix represents a linear transformation between vector spaces
of the same dimension, where the domain and codomain are the same. When a square matrix
is invertible, its linear transformation is bijective, meaning that it has a unique inverse
transformation.









Let's consider why non-square matrices cannot have inverses:


1. If a matrix A has more rows than columns (m > n), i.e., a __tall matrix,__ the linear
transformation it represents is __from a lower-dimensional space to a higher-dimensional
space.__ In this case, the transformation is generally not surjective (onto), as there are
output vectors in the higher-dimensional space that have no corresponding input vector.
Consequently, there is no inverse transformation that can map every output vector back
to an input vector.




2. If a matrix A has more columns than rows (m < n), i.e., a __wide matrix,__ the linear
transformation it represents is __from a higher-dimensional space to a lower-dimensional
space (dimension reduction).__ In this case, the transformation is generally not injective
(one-to-one), as multiple input vectors can map to the same output vector. Consequently,
there is no unique inverse transformation that can map each output vector back to a
unique input vector.



Again, the inverse of a matrix is possible only for square matrices because these matrices
represent linear transformations between vector spaces of the same dimension. Only in these
cases can a matrix potentially satisfy the conditions of being a bijective transformation, i.e.,
both injective and surjective, which allows the existence of a unique inverse transformation.

However, not all square matrices have inverses; only those that are non-singular (with a non-
zero determinant) have an inverse.

### __conclusion :__

Think of a non-square matrix like a rectangle instead of a square. Just like a rectangle doesn't have the same length and width, a non-square matrix doesn't have the same number of rows and columns. Because of this difference, it's like having too many or too few directions, so the matrix can't be easily reversed to get back to where you started. In other words, you can't find an inverse for a non-square matrix because it doesn't match up neatly.

### Hadamard product

The Hadamard product, also known as the element-wise product or Schur
product, is a binary operation that takes two matrices of the same dimensions
and produces a third matrix where each element is the product of the
corresponding elements of the input matrices. Specifically, given two matrices
A and B of the same size m  n, their Hadamard product C is also an m  n
matrix, where each element is defined as:

C[i, j] = A[i, j] * B[i, j]

for all i = 1, ..., m and j = 1, ..., n.
