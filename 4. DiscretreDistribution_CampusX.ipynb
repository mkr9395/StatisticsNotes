## Uniform Distribution

In probability theory and statistics, a uniform distribution is a probability distribution
where __all outcomes are equally likely within a given range.__ 

This means that if you were to
select a random value from this range, any value would be as likely as any other value.

â€¢ Examples :

- a. The height of a person randomly selected from a group of individuals whose heights
range from 5'6" to 6'0" would follow a continuous uniform distribution.


- b. The time it takes for a machine to produce a product, where the production time ranges from 5 to 10 minutes, would follow a continuous uniform distribution.


- c. The distance that a randomly selected car travels on a tank of gas, where the distance ranges from 300 to 400 miles, would follow a continuous uniform distribution.


- d. The weight of a randomly selected apple from a basket of apples that weighs between 100 and 200 grams, would follow a continuous uniform distribution.

### <span class="mark">Application of Uniform Distribution in Machine learning and Data Science</span>

Uniform distribution is applied in machine learning and data science in the following ways:

1. **Random Initialization**: It's used to randomly initialize parameters in algorithms like neural networks and k-means clustering, ensuring all values in a range have an equal chance of selection. eg ; generating random numbers


2. **Sampling**: Uniform distribution helps in random sampling, ensuring fair representation of all classes in datasets. eg; train-test split


3. **Data Augmentation**: It's used to generate new data points similar to the original data, effectively increasing dataset size.


4. **Hyperparameter Tuning**: Uniform distributions are employed in hyperparameter tuning to sample and explore the hyperparameter space effectively.

## Bernoulli Distribution

Bernoulli distribution is a probability distribution that models a __binary outcome,__ where the
outcome can be either success (represented by the value 1) or failure (represented by the
value 0)


Conditions for Bernoulli Distribution:
- 1 Trial
- 2 Possible outcomes
- eg: tossing a coin, marking True or False
<br></br>
We often have to assign which outcome is 0 and which outcome is 1. 

$$0\leq\;p\;\leq1$$

$$q\;=\;1\;-\;p$$



After doing so, we can calculate the expected value i.e $E(X) = p\;or\;E(X) = 1-p$
<br></br>
Conventionally, we also assign a value of 1 to the event with the probability equal to P and 0 to 1-p. So, E(X) expresses the likelihood of the favoured Event. Since we only have one trial and a favorite event. We expect that outcome to occur. i.e $E(X) = p$

$$\text{Variance of bernoulli distribution is always} = p(1-p)$$

__The Bernoulli distribution is characterized by a single parameter, which is the probability of
success, denoted by p.__

The Bernoulli distribution is commonly used in machine learning for modelling
binary outcomes, such as whether a customer will make a purchase or not,
whether an email is spam or not, or whether a patient will have a certain disease
or not.

### As Bernoulli has discrete outcomes so it will have Probability Mass function(PMF)

#### PMF formula for Bernoulli:

$$P(X = x) = 
\begin{cases} 
1-p & \text{if } x = 0 \\
p & \text{if } x = 1 
\end{cases}$$

```
P(X = 0) = 1 - p
P(X = 1) = p
```

#### CDF (Cumulative Distribution Function) for Bernoulli :

$$
F(x) = 
\begin{cases} 
0 & \text{if } x < 0 \\
1 - p & \text{if } 0 \leq x < 1 \\
1 & \text{if } x \geq 1
\end{cases}
$$

```
F(x) = 0   if x < 0
F(x) = 1 - p   if 0 <= x < 1
F(x) = 1   if x >= 1
```

## Binomial Distribution

Binomial distribution is a probability distribution that describes the number of
successes in a fixed number of independent Bernoulli trials with two possible
outcomes (often called "success" and "failure"), where the probability of success
is constant for each trial. 

The binomial distribution is characterized by two
parameters: the number of trials n and the probability of success p.

- Binomial Distribution is a sequence of identical Bernoulli events.


- It is repeated for fixed number of times.


- The trials are independent and have 2 mutually exclusive outcomes either success or failure.


- The probability of success is same for all trials

$$It\;is\;denoted\;by "B"$$ 

$$B\;(n,p)$$ 

$$X \sim B\;(10,0.6)$$

We read the following statement as variable X follows a binomial distribution with 10 trials and a likelihood of 0.6 to succeed on each individual trial i.e it's not biased


### **Probability distribution function :**

>$$p(y) = {n \choose x}*p^x*(1-p)^{n-x}$$
 OR
 
>$$p(y)\;=\;{}^nC_x*p^x*(1-p)^{n-x}$$


Here $n\;\longrightarrow$ is number of trials and $p \longrightarrow$ is success probability for each trials.
<br></br>

#### NOTE : We can express a Bernoulli distribution as a binomial distribution with a single trial.

$$Bern(p)=Bern(1,p)$$

___eg: there is a 60% i.e 0.6 chance of general motor share to increase.___

___So to calculate 3 increase in 5 days:___
    
$y \rightarrow 3 , $ $n \rightarrow 5 , $ $p \rightarrow 0.6$
    
$$p(x) = 5^C_3 * 0.6^3 * 0.4^2 \;\; = \;0.3456\;or\;34.56 \% $$

There is 34.56% chance of increase in stock value 3 out of 5 days.

### Examples of Binomial Distribution : 


**Example 1: Coin Flips**

- __Situation: You flip a fair coin 5 times, and you want to know the probability of getting exactly 3 heads.__


- Parameters: Number of trials (n) = 5, Probability of success (p) = 0.5 (because a fair coin has a 50% chance of landing heads).


- Calculation: You can use the Binomial probability formula:

  $$P(X = x) = \binom{n}{x} \cdot p^x \cdot (1-p)^{n-x}$$

  - $(x)$ is the number of successes you want (3 in this case).
  
  
  - $\binom{n}{x}$ represents the number of ways to choose $(x)$ successes out of $(n)$ trials, which is calculated as $\binom{5}{3} = 10$.
  
  
  - Plug in the values: $(P(X = 3) = 10 \cdot (0.5)^3 \cdot (0.5)^2 = 0.3125)$.

#### using `stats.binom.pmf (k, n, p)` :

````python
from scipy import stats
````

````python
n = 5  # Number of trials
p = 0.5  # Probability of success (fair coin)

# Number of heads you want (k)
k = 3
````

````python
stats.binom.pmf(k, n, p)
````

````python
print(f'probability of getting exactly 3 heads {(stats.binom.pmf(k, n, p))*100:.3f}%')
````

**Example 2: Exam Questions**

- __Situation: In a multiple-choice exam with 10 questions, each with 4 choices, you randomly guess the answers. You want to find the probability of getting exactly 3 correct answers.__


- Parameters: Number of trials (n) = 10 (number of questions), Probability of success (p) = 0.25 (because there is a 1 in 4 chance of guessing correctly).


- Calculation: Again, use the Binomial probability formula:

  $$P(X = x) = \binom{n}{x} \cdot p^x \cdot (1-p)^{n-x}$$

  
  - x is the number of correct answers you want (3 in this case).
  
  
  - Calculate $\binom{10}{3} = 120$ ways to choose 3 correct answers out of 10.
  
  
  - Plug in the values: $P(X = 3) = 120 \cdot (0.25)^3 \cdot (0.75)^7 \approx 0.250$

````python
n = 10  # Number of trials
p = 0.25  # Probability of success

# Number of correct answers (k)
k = 3
````

````python
stats.binom.pmf(k, n, p)
````

````python
print(f'probability of getting exactly 3 correct answers {(stats.binom.pmf(k, n, p))*100:.3f}%')
````

**Example 3: Defective Products**

- __Situation: In a production line, each product has a 10% chance of being defective. You want to find the probability that in a batch of 20 products, exactly 2 are defective.__


- Parameters: Number of trials (\(n\)) = 20, Probability of success (\(p\)) = 0.10 (because a 10% chance of being defective).


- Calculation: Use the Binomial probability formula as before:

  $$P(X = x) = \binom{n}{x} \cdot p^x \cdot (1-p)^{n-x}$$

  
  - (x) is the number of defective products you want (2 in this case).
  
  
  - Calculate $(\binom{20}{2} = 190)$ ways to choose 2 defective products out of 20.
  
  
  - Plug in the values: $P(X = 2) = 190 \cdot (0.10)^2 \cdot (0.90)^{18} \approx 0.285$.

````python
n = 20  # Number of trials
p = 0.10  # Probability of defect

# Number of defectives (k)
k = 2
````

````python
stats.binom.pmf(k,n,p)
````

````python
print(f'probability of exactly 2 are defective {(stats.binom.pmf(k, n, p))*100:.3f}%')
````

### Criteria for Binomial Distribution

- 1. The process consists of n trials


- 2. Only 2 exclusive outcomes are possible, a success and a failure.


- 3. P(success) = p and P(failure) = 1-p and it is fixed from trial to trial


- 4. The trials are independent.

#### <span class="mark">Applications of  Binomial Distribution:</span>

1. ___Binary classification problems:___ In binary classification problems, we often model
the probability of an event happening as a binomial distribution. For example, in a
spam detection system, we may model the probability of an email being spam or
not spam using a binomial distribution.


2. ___Hypothesis testing:___ In statistical hypothesis testing, we use the binomial
distribution to calculate the probability of observing a certain number of
successes in a given number of trials, assuming a null hypothesis is true. This can
be used to make decisions about whether a certain hypothesis is supported by
the data or not.


3. ___Logistic regression:___ Logistic regression is a popular machine learning algorithm
used for classification problems. It models the probability of an event happening
as a logistic function of the input variables. Since the logistic function can be
viewed as a transformation of a linear combination of inputs, the output of
logistic regression can be thought of as a binomial distribution.


4. ___A/B testing:___ A/B testing is a common technique used to compare two different
versions of a product, web page, or marketing campaign. In A/B testing, we
randomly assign individuals to one of two groups and compare the outcomes of interest between the groups. Since the outcomes are often binary (e.g., click- through rate or conversion rate), the binomial distribution can be used to model the distribution of outcomes and test for differences between the groups.

> Q- In a survey, it was found that 85% of the household have high speed internet. If you take sample of 18 Households, what is the probability that ___atleast___ 15 will have high speed internet? 


````python
n = 18  # Number of trials
p = 0.85  # Probability of defect

# Number of at least 15 (k)
k = 15
````

#### for x > 15 :

````python
1- stats.binom.cdf(k,n,p)
````

#### now for 15 : 

````python
stats.binom.pmf(k,n,p)
````

#### now for x$\geq$ 15 :  x > 15  + x = 15

````python
print(f'probability that atleast 15 will have high speed internet {((1- stats.binom.cdf(k,n,p))+(stats.binom.pmf(k,n,p)))*100:.3f}%')
````

#### OR

````python
import scipy.stats as stats

# Define the parameters for the binomial distribution
n = 18  # Number of trials (households in the sample)
p = 0.85  # Probability of success (having high-speed internet)

# Initialize the cumulative probability
cumulative_probability = 0

# Calculate the cumulative probability for X â‰¥ 15
for k in range(15, n + 1):
    cumulative_probability += stats.binom.pmf(k, n, p)

print(f'The probability that at least 15 out of 18 households have high-speed internet is: {cumulative_probability:.4f}')
````
