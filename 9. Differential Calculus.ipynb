### What is differentiation

Differentiation is the process of finding the __derivative of a function.__


The derivative of a
function represents the __instantaneous rate of change__ of the function with respect to its
variable, typically denoted as 'x'.

#### interval rate of change between x and (x + $\Delta$ x):

$$\frac{f(x+\Delta x)-f(x)}{\Delta x}$$

### but we want instantaneous rate of change:

- we decrease the interval make and it very very small i.e $\Delta x$ tending to zero $\rightarrow$ 0.000000001 


- $\Delta x$ becomes dx

### Why instantaneous

__Differentiation is instantaneous because it measures the rate at which a function changes at a specific point.__



In calculus, the derivative of a function gives you the slope of the tangent line to the curve representing that function at a particular point. Since the tangent line at a single point represents the best linear approximation of the function's behavior near that point, the derivative provides an instantaneous rate of change at that specific location. 


__This is why differentiation is considered to be instantaneous  it captures how the function behaves at an infinitesimally small scale around a chosen point.__

### Why calculating derivatives are useful:


1. __for rate of change :__ at very small interval of change i.e instaneous change, derivative are considered as slope to the point.


2. __minima and maxima :__ but calculating the derivative / slope $(\frac{dy}{dx})$ and making the slope equation  = 0 , we can calculate those values of x which will maxima or minima of the curve.

## Derivatives

### Common Derivatives:

### 1. Constant Rule:
 $$\frac{d}{dx}(c) = 0 \text{, where (c) is a constant}$$

### 2. Power Rule:
   $$\frac{d}{dx}\;x^n = nx^{n-1} \text{, where (n) is a real number}$$

### 3. Sum Rule:
   $$\frac{d}{dx}\;(f(x) + g(x)) = \frac{d}{dx}f(x) + \frac{d}{dx}g(x)$$

### 4. Product Rule:
    
$$\frac{d}{dx} (f(x) \cdot g(x)) = f'(x) \cdot g(x) + f(x) \cdot g'(x)$$

### 5. Quotient Rule:
$$\frac{d}{dx} \Big(\frac{f(x)}{g(x)}\Big) = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{[g(x)]^2}$$

### 6. Chain Rule:
$$\frac{d}{dx} f\big(g(x)\big) = f'(g(x)) \cdot g'(x)$$

#### eg 1 : $$y = sin x^2$$



#### eg 2 : $$y=sin\big(log(x^2)\big)$$

### 7. Exponential Rule:
   $$\frac{d}{dx}e^{u} = e^{u} \cdot \frac{du}{dx}$$

### 8. Logarithmic Rule:
   $$\frac{d}{dx}\;\ln(u) = \frac{1}{u} \cdot \frac{du}{dx}$$

## Partial Differentiation : 

The partial derivative of a function f(x, y) with respect to x is denoted as:

$$
\frac{\partial f}{\partial x} = \lim_{{\Delta x} \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x}
$$

Similarly, the partial derivative of f(x, y) with respect to y is denoted as:

$$
\frac{\partial f}{\partial y} = \lim_{{\Delta y} \to 0} \frac{f(x, y + \Delta y) - f(x, y)}{\Delta y}
$$


### NOTE : 

$$\text{slope of x-direction}\;\big(\frac{\partial z}{\partial x}\big)\;and\;\text{slope of y-direction}\;\big(\frac{\partial z}{\partial y}\big)\;\text{will always be perpendicular to each other}$$

#### eg :

$$z=x^2+y^2$$

## Higher order Derivatives:

- __Second Order Derivative:__<br></br>

    - Instantaneous rate of change of slope (slope  already in itself is instantaneous rate of change). So we are doing differentitation twice.<br></br>
    
    - The second order derivative of a function measures how the rate of change of its first order derivative changes with respect to the variable. It indicates the curvature or concavity of the function.<br></br>
    - eg : Imagine a car's speed changing. The second order derivative is like looking at how the car's acceleration is itself changing  if it's speeding up or slowing down.

$$
f''(x) = \frac{d^2f}{dx^2} = \frac{d}{dx}\left(\frac{df}{dx}\right)
$$

- __Third Order Derivative:__<br></br> 
    - The third order derivative of a function describes how the rate of change of its second order derivative changes. It helps analyze features such as inflection points and provides insights into the behavior of curvature changes.<br></br> 
    - Continuing with the car analogy, the third order derivative is like examining how the car's acceleration is changing its change  whether it's accelerating faster or slowing down its acceleration.


$$
f'''(x) = \frac{d^3f}{dx^3} = \frac{d}{dx}\left(\frac{d^2f}{dx^2}\right)
$$

### Why do need higher order derivative?

$$\frac{d^2y}{dx^2}\;\Longrightarrow\;2^{nd} \text{order derivative gives instantaneous rate of change of slope}$$ 

- if a slope = 0 so $\frac{dy}{dx}$ will be a maxima or minima.


- but when differentiate it again then the second order $\frac{d^2y}{dx^2}$ will tell exactly it is a maxima or a minima.

- __slope = -ve then maxima__


- __slope = +ve then minima__
