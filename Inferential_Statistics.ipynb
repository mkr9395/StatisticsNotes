### Z-Score
#### The Standard Normal Distribution
<br></br>
Standardization is the process of transforming this variable to one with a mean of zero$(\mu=0)$ and a standard deviation$(\sigma = 1)$ of one.
The Standard Normal distribution is a particular case of the Normal distribution It has a mean of 0 and a standard deviation of 1.

Every Normal distribution can be using the standardization formula:

$$Formula\;for\;calculating\;z-score$$

$$z\;=\;\frac{x-\mu}{\sigma}$$
$$where\; z\;is\;the\;z-score$$
$$where\; x\;is\;the\;original\;data\;points\;variable$$
$$N\;\sim\;(0,1)$$

>so if we shift the mean by $\mu$ and the standard deviation by $\sigma$ for any normal distribution, we will arrive at the standard normal distribution.

    Why standardize?
    - compare different normally distributed datasets
    - detect normality
    - detect outliers
    - create confidence intervals
    - test hypotheses
    - perform regression analysis
    
<br></br>

>> **NOTE:  Through Z-Score Gaussian\Normal Distribution is converted into Standard Normal Distribution**


>**Rationale of the formula for standardization**
We want to transform a random variable from $N \sim\;(\mu,\sigma^2)\;to\;N \sim (0,1)$. Subtracting the mean from all observations would cause a transformation from $N \sim\;(\mu,\sigma^2)\;to\;N \sim (0,\sigma^2)$.,moving the graph to the origin. Subsequently, dividing all observations by the standard deviation would cause a transformation from $N \sim\;(\mu,\sigma^2)\;to\;N \sim (0,1)$ standardizing the peak and the tails of the graph

>>> A random variable belongs to standard normal distribution when mean is 0 and standard deviation is 1. 
This process is known as **Standardization.** i.e. bringing all features under one scale through Z-score

### Formula for Z-Score:

Used to calculate standard deviation of a point from it's mean. It lies in which Standard Deviation and left or right side.

$$ Z = \frac{x_i - \mu}{\sigma}$$

$$where\;\sigma\;is\;Standard\;Deviation$$

**Z-Score is used to converts Normal Distribution into Standard-Normal Distribution**

### CENTRAL LIMIT THEOREM
<BR></BR>

**Sampling Distribution of mean :** When we are referring to a distribution form by samples of means, we use the term a sampling distribution.

$$CLT$$
No matter the distribution of the population, binomial, uniform, exponential or another one, the sampling distribution of the mean will approximate a normal distribution. Not only that, but it's mean is the same as the population mean.

It states that no matter the underlying distribution of the dataset, the sampling distribution of the means would approximate a **normal \ gaussian distribution**. Moreover, the mean of the sampling distribution would be **equal to the mean of the original distribution and the variance would be n times smaller**, where n is the size of the samples. The CLT applies whenever we have a sum or an average of many variables (e g sum of rolled numbers when rolling dice)

No matter the underlying distribution, the distribution of $\bar x_1, \bar x_2,\bar x_3,....\bar x_k$ would tend to

$$N\sim\;(\mu,\;\frac{\sigma^2}{n})$$
$$where \frac{\sigma^2}{n}\;is\;standard\;deviation$$
$$The\;more\;samples,\;the\;closer\;to\;Normal\;(k\rightarrow \infty)$$
$$The\;bigger\;the\;samples,\;the\;closer\;to\;Normal\;(n\rightarrow \infty)$$

Since the sample size is in the denominator, the bigger the sample size, the lower the variance or in other words, the closer the approximation we get.
Usually for CLT to apply, we need a sample size of at least 30 observations.
    
    Why is CLT Useful?
>- The CLT allows us to assume normality for many different variables. That is very useful for confidence intervals, hypothesis testing, and regression analysis. In fact, the Normal distribution is so predominantly observed around us due to the fact that following the CLT, many variables converge to Normal.
    - They approximate a wide variety of random variables.
    - Distribution of sample means with large enough sample sizes could be approximated to normal.

### Standard Error:

The standard error is the standard deviation of the distribution form by the sample means, in other words, the standard deviation of the sampling distribution.

It's variance sigma squared divided by N, therefore, the standard deviation is sigma divided by the square root of N.

$$Standard\;Error\;(of\;sampling\;distribution)\;=\; \sqrt \frac{\sigma^2}{n}\;\;=\;\;\frac{\sigma}{\sqrt n}$$

The standard error shows variability, it is the variability of the means of the different samples we extracted.

Like a standard deviation, the standard error shows variability in this case, it is the variability of the means of the different samples we extracted.

Standard error decreases as the sample size increases, this makes sense as bigger samples give a better approximation of the population.

**Estimate and Estimators**

<br></br>
**Estimators :**  an estimator is a mathematical function that approximates a population parameter depending only on sample information.

Estimators have two important properties:
- **Bias :** The expected value of an unbiased estimator is the population parameter The bias in this case is 0. If the expected value of an estimator is (parameter b), then the bias is b.
- **Effieciency :** The most efficient estimator is the one with the smallest variance.

<br></br>
**Estimates :** An estimate is the output that you get from the estimator (when you apply the formula).

There are two types of estimates, point estimates and confidence interval estimates.
- **Point Estimates :** The value of any statistic that estimates the value of a parameter. 
                    : A single value. eg: 1, 2, 5.5, 0.32
- **Confidence Intervals :** An Interval. eg: (1,5), (12,33), (-0.71,0.11)

Confidence intervals are much more precise than point estimates. That is why they are preferred when making inferences

### Confidence Interval 

A confidence interval is an interval within which we are confident (with a certain percentage of confidence) the population parameter will fall. We build the confidence interval around the point estimate. 

>*Point Estimates* is the mid-point of an interval.

$(1-\alpha)$ is the level of confidence. We are $(1-\alpha)$ * 100 % confident that the population parameter will fall in the specified interval. Confidence level = 95% then $\alpha$ = 5%

$$\alpha\;is\;significance\;value$$

**General formula:**  $$[\bar \chi - ME, \bar \chi + ME] \;\;where\;ME\;is\;Margin\;of\;Error.$$
$$where\; \bar \chi \; is\;sample\;mean\;or\;Point\;Estimate$$

$$Confidence\;Interval\;=\;Point\;Estimate\;\pm\;Margin\;of\;Error$$

### Student's T Distribution
$$Degree\;of\;Freedom\;=\;(n-1)$$
<br></br>
Visually, the students-t distribution looks much like a normal distribution, but generally has fatter tails. Fatter tales, allows for a higher dispersion of variables and there is more uncertainty. This means the probability of values being far away from the mean is bigger. For big enough samples, the Students T distribution coincides with the Normal distribution. In the same way that the z- statistic is related to the standard normal distribution, the T- statistic is related to the students t distribution.

The Students T distribution is used predominantly for creating confidence intervals and testing hypotheses with normally distributed populations when the sample sizes are small. It is particularly useful when we dont have enough information or it is too costly to obtain it.
<br></br>

All else equal, the Students T distribution has fatter tails than the Normal distribution and a lower peak This is to reflect the higher level of uncertainty, caused by the small sample size.

#### Margin of Error (ME)

<br></br>
Getting a smaller margin of error means that the confidence interval would be narrower as we want a better prediction, it is in our interest to have the narrowest possible confidence interval.

The best part is that we can control the margin of error.
 
 The true population mean falls in the interval defined by the $sample\; mean \;(\bar \chi)\;\pm\;the\;margin\;of\;error.$
 
 where $$margin\;of\;error\;=\;reliability\;factor\;*\; \frac{standard\;deviation}{\sqrt {sample\;size}}$$
 
 $$reliability\;factor\;=\; t-score\; \left\lceil t_{n-1},\frac{_\alpha}{_2}\right\rceil\;\;or\;\;z-score \left\lfloor z_\left(\frac{\alpha}{2}\right)\right\rfloor$$
 
 $$Standard\;Error\longrightarrow\frac{standard\;deviation}{\sqrt {sample\;size}}$$
 
<br></br>

<br></br>
A lower standard deviation$(\sigma)$ means that the data set is more concentrated around the mean. So we have a better chance to get it right.
<br></br>
Higher sample sizes$(n)$ will decrease the margin of error. The more observations you have in your sample, the more certain you are in the prediction.

        - A higher statistic increases the Margin of Error (ME). 
        - A higher standard deviation increases the  ME. 
        - A higher sample size decreases the ME.

## HYPOTHESIS TESTING

<br></br>
- **A hypothesis is "an idea that can be tested"**
- It is a supposition or proposed explanation made on the basis of limited evidence as a starting point for further investigation.
<br></br>
$Null\;Hypothesis\;(H_0)$ 
- The null hypothesis is the hypothesis to be tested.
- It is the status quo Everything which was believed until now that we are contesting with our test
- The concept of the null is similar to innocent until proven guilty We assume innocence until we have enough evidence
to prove that a suspect is guilty.
- Conventionally, the equality sign is always included in the null.

<br></br>
$Alternative\;Hypothesis\;(H_1\;or\;H_A)$
- The alternative hypothesis is the change or innovation that is contesting the status quo.
- Usually the alternative is our own opinion The idea is the following:
    - If the null is the status quo (i e what is generally believed), then the act of performing a test, shows we have doubts about the truthfulness of the null More often than not the researchers opinion is contained in the alternative hypothesis.

<br></br>
**In statistics, the null hypothesis is the statement we are trying to reject. Therefore, the null hypothesis is the present state of affairs, while the alternative is our personal opinion.**

eg: 

1. So, for the data scientist salary example, the null would be the mean data scientist salary is $\$ 113,000$ Then we will try to reject the null with a statistical test So, usually, your personal opinion (e g data scientists don't earn exactly that much) is the alternative hypothesis.
2. Our friend Paul told us that the mean salary is > $\$ 125,000$ (status quo, null) Our opinion is that he may be wrong, so we are testing that Therefore, the alternative is the mean data scientist salary is lower or equal to $\$125,000$

It is the probability of rejecting the null hypothesis if it is true, so the probability of making this error.

### Z-TEST
$$Standardization\;formula$$
$$Z-test:\;Z = \frac{\bar \chi - \mu_0}{\frac{s}{\sqrt n}}$$

$$where\;"\bar\chi"\;is\;sample\;mean,\;\;\; "\mu_0"\;is\;hypothesized\;mean$$

$$\frac{s}{\sqrt n}\; is\;Standard\;Error$$

We are standardizing or scaling the sample mean we got. If the sample mean is close enough to the hypothesized mean, then Z-test will be close to zero. Otherwise, it will be far away from it.

Naturally, if the sample mean is exactly equal to the hypothesis mean, $\rightarrow\bar\chi = \mu_0\;\;$ Z will be zero $\rightarrow Z=0$. *In all these cases, we would accept the null hypothesis.*

>When we calculate Z, we will get a value, if this value falls into the middle part, then we cannot reject the null. If the Z-value falls outside in the shaded region, then we reject the null hypothesis.That is why the shaded part is called rejection region. The area that is cut off actually depends on the significance level.
so if level of significance $\alpha\;=\;0.05$ then $\frac{\alpha}{2}\;=\;0.025$ so according to z-table $z = 1.96$

>so +1.96 on the right side and -1.96 on the left side. If the value we get Z from the test is lower than -1.96  or higher than 1.96, we will reject the null hypothesis. Otherwise we will accept it.

>**NOTE :**
<br></br>
Level of Significance $\longrightarrow "\alpha"$
<br></br>
Level of Confidence $\longrightarrow "1-\alpha"$
<br></br>
Both are opposite to each other i.e. $(\alpha)\;+\;(1-\alpha)\;=\;1$

>**NOTE**
<br></br>
Level of significance $\longrightarrow$ For one tailed test we take '$\alpha$' but  
Level of significance $\longrightarrow$ For two-tailed test we consider '$\frac{\alpha}{2}$'

### p-value
<br></br>
- p-value is the smallest level of significance at which we can still reject the null hypothesis, given the observed sample statistics.
<br></br>
- When we are testing a hypothesis, we always strive for those *'three zeros after the dot'* This indicates that we reject the null at all significance levels.
<br></br>
- 0.05 is often the *cut off line*. If our p-value is higher than 0.05 we would normally accept the null hypothesis (equivalent to testing at 5% significance level). **If the p-value (significance-value) is lower than or equal to 0.05, we would reject the null.**

<br></br>
**Rule:**
We should reject the null hypothesis, if:
$$p-value\;<\;\alpha\;(significance\;level)$$

- test at 90% : 0.00001 < 0.1 $\rightarrow$ REJECT THE NULL HYPOTHESIS
- test at 95% : 0.00001 < 0.05 $\rightarrow$ REJECT THE NULL HYPOTHESIS
- test at 99% : 0.00001 < 0.01 $\rightarrow$ REJECT THE NULL HYPOTHESIS

#### Statistical errors (Type I Error and Type II Error)

There are two types of errors we can make while testing:
1. **Type I error (False positive)**
    - *wrongly invited her*
    - When you reject a true null hypothesis. 
    - The probability of making this error is $\alpha$, the level of significance.
    - Since you, the researcher, choose the Alpha, the responsibility for making this error lies solely on you.
<br></br>
2. **Type II Error (False negative)**
    - *missed your chance*
    - when you accept a false null hypothesis.
    - The probability of making this error is $\beta$.
    - $\beta$ depends on sample size $(n)$ and variance $(\sigma)$
    - So if your topic is difficult to test due to hard sampling or the effect you are looking for is almost negligible, it is more likely to make this type of error.
    - **Power of the test:** Probability of Rejecting a null hypothes is False. $(1-\beta)$
        - Power of test is increased by increasing the sample size.
