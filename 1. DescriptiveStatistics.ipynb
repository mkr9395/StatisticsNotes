# STATISTICS

Statistics is a branch of mathematics that involves collecting,
analysing, interpreting, and presenting data. It provides tools and
methods to understand and make sense of large amounts of data
and to draw conclusions and make decisions based on the data.

### TYPES of STATISTICS : 

### 1. Descriptive statistics


- Descriptive statistics are used to summarize and describe data in a clear and meaningful way. 


- They provide a snapshot of the main characteristics of a dataset, such as central tendency (mean, median, mode), dispersion (range, variance, standard deviation), and the shape of the data distribution (e.g., histogram or frequency table). 


- Descriptive statistics help us understand the basic features of data without making any inferences or drawing conclusions beyond the dataset itself.



### 2. Inferential statistics

- Inferential statistics, on the other hand, go beyond the data at hand to make predictions, draw conclusions, or test hypotheses about a larger population based on a sample from that population. 



- They involve using sample data to make inferences or generalizations about a broader group. Common techniques in inferential statistics include hypothesis testing, confidence intervals, and regression analysis. 



- Inferential statistics help us make decisions or reach conclusions about a population based on limited data from a sample.

## Measures of Central Tendency

A measure of central tendency is a statistical measure that represents a typical or
central value for a dataset. It provides a summary of the data by identifying a
single value that is most representative of the dataset as a whole.

### 1. Mean:  

The Mean is the most widely spread measure of central tendency. It is the simple average of the dataset.

* Population mean is denoted by $\mu$ 
* Sample mean is denoted by $\bar \chi$

>Note $\rightarrow$ Downside of Mean: It is easily affected by outliers

__Formula of Population Mean__ $(\mu)$ = $$\frac{x_1+x_2+x_3+x_4...+x_{N-1}+x_N}{N}$$

$$or$$
$$\frac{1}{N} * \sum_{i=1}^N x_i $$


__Formula of Sample Mean__ $(\bar x)$ = 

$$\bar x = \frac{1}{n} * \sum_{i=1}^n x_i$$

### 2. Median: 
The median is the midpoint of the **ordered dataset**. It is not as popular as the mean, but is often used in academia and data science. That is since it is not affected by outliers. 


In an ordered dataset the formulas for calculating the median of a dataset are as follows:

1. **Odd Number of Observations (N is odd):**

   $\text{Median (M) = Middle Value}$



2. **Even Number of Observations (N is even):**

   $\text{Median (M)} = \frac{\text{(Middle Value 1 + Middle Value 2)}}{2}$

In both cases, we first need to arrange the data in ascending order, and then apply the appropriate formula based on whether the number of observations is odd or even to find the median. 



___For an odd number of observations, the median is the middle value, and for an even number of observations, the median is the average of the two middle values.___

### 4. Weighted Mean




A weighted mean, also known as a weighted average, is a type of average that takes into account the relative importance or weights of different values in a dataset. It is calculated by multiplying each value by its respective weight, summing these products, and then dividing by the sum of the weights. 


___It is used to calculate a mean when
the values in the dataset have different importance or frequency.___



The formula for the weighted mean is:

$$\text{Weighted Mean (} \overline{x} \text{)} = \frac{\sum (x_i \cdot w_i)}{\sum w_i}$$



Where:
- $ \overline{x} $ is the weighted mean.
- $ x_i $ represents each value in the dataset.
- $ w_i $ represents the weight associated with each value.
- $ \sum $ denotes summation, meaning you should add up the products of $ x_i $ and $ w_i $ for all values.

Here's an example to illustrate how to calculate a weighted mean:

Suppose you want to calculate the weighted mean score of a student's grades in three courses, where the courses have different credit hours (weights):

- Grade in Course A: 90 (weight = 3 credit hours)
- Grade in Course B: 85 (weight = 2 credit hours)
- Grade in Course C: 78 (weight = 4 credit hours)

To find the weighted mean:

Weighted Mean (XÌ„) = [(90 * 3) + (85 * 2) + (78 * 4)] / (3 + 2 + 4)

Weighted Mean (XÌ„) = (270 + 170 + 312) / 9

Weighted Mean (XÌ„) = 752 / 9 â‰ˆ 83.56 (rounded to two decimal places)

So, the weighted mean score for the student is approximately 83.56, taking into account the credit hours (weights) of each course. This means that the student's performance in Course C, which has more credit hours, has a greater influence on the weighted mean than the other courses.

## <span class="mark">Why we use (n-1) in sample?</span>

___Dividing by  "N" underestimates the population variance___

- Here $(n-1)$ is  **Bessel's Correction** which is used for ___unbaised estimation___ of population when calculating sample variance.

Here's a more detailed explanation:

1. __Bessel's Correction:__ 
    - The (n-1) corrects for the fact that when you use the sample mean to estimate the population mean, you have effectively used one piece of information from your data (the mean itself).This reduces the variability in your calculation.<br></br>
    - By dividing by (n-1) instead of n, you give each data point more "freedom" to vary and provide a more unbiased estimate of the population variance or standard deviation.


2. __Unbiased Estimation:__ 
    - Using (n-1) instead of n helps to produce unbiased estimates of the population parameters.<br></br> 
    - An unbiased estimator is one that, on average, gives you an estimate that is equal to the true population parameter. Dividing by (n-1) in the sample variance and standard deviation calculations helps achieve this property.

In summary, using (n-1) in statistics calculations for sample variance and sample standard deviation is a standard practice to correct for bias and provide a more accurate estimate of the population parameters when working with sample data. ___It accounts for the loss of one degree of freedom due to using the sample mean in the calculations.___

__NOTE : Population Variance will be **less than** Sample variance as sample variance has (n-1) in denominator  resulting in bigger numbers__

- The primary reason for using (n-1) instead of n is to account for the __degrees of freedom in the sample data.__<br></br>

    - Degrees of freedom represent the number of values in the final calculation that are free to vary.<br></br>
       
    - When you calculate the mean (average) of a sample, you use n (the number of data points in the sample) in the denominator because all the data points are used in the calculation, and none of them are free to vary.<br></br> 
    
    - However, when calculating the sample variance or sample standard deviation, you use (n-1) in the denominator to account for the fact that one degree of freedom is lost when you use the sample mean to estimate the population mean.

> The sample mean is supposed to be equal or closest to population mean. But you have the freedom to choose the samples randomly.<br></br>
> So, to keep the sample mean equal to the population mean you can change (n-1) numbers of sample.
Suppose, there are 5 samples, you can change only (5-1) = 4 sample... Because the only sample you are not changing will keep the sample mean closest or equal to population mean.

 The CV allows to compare the variability of different data sets, even if they have different units or scales. A higher CV indicates greater relative variability, while a lower CV suggests lower relative variability.

### Difference between coefficient of variance and variance/std dev:

### Spearman's Rank Correlation $(\rho)$

1. **Definition:** Spearman correlation, also known as Spearman's rank correlation coefficient, assesses the strength and direction of the monotonic relationship between two variables. ___It does not assume linearity and is appropriate for both continuous and ordinal data.___


2. **Assumption:** Spearman correlation does not assume that the data follows a specific distribution or have a linear relationship. ___It is less sensitive to outliers compared to Pearson correlation because it relies on rank ordering of data rather than the actual values.___



3. **Range:** The Spearman correlation coefficient also ranges from -1 to 1, with similar interpretations as Pearson correlation:
   - A correlation of 1 indicates a perfect positive monotonic relationship.
   - A correlation of -1 indicates a perfect negative monotonic relationship.
   - A correlation of 0 suggests no monotonic relationship (but doesn't rule out other types of relationships).



4. **Interpretation:** Spearman correlation assesses the strength and direction of a monotonic relationship. Monotonicity means that as one variable increases, the other either consistently increases or decreases, without necessarily being linear. It is more robust to outliers and can capture nonlinear relationships



It captures the more non-linear properties of relation between x and y.

$$rho\; = \; \frac{\text{Cov(Rank of x, Rank of y)}}{\text{Std dev(rank of x)}*\text{Std dev(rank of y)}}$$

$$rho\; = \; \frac{Cov(rg_X,rg_Y)}{Stdev(rg_X)\;*\;Stdev(rg_Y)}$$



##### here Correlation is 0.88 instead of 1 but value of y is increasing with increase of x.
##### for the same reason Spearman Correlation was founded.

> __meaning of monotonic -__ the general trend or pattern between two variables where, as one variable increases, the other either consistently increases or consistently decreases (__but not necessarily at a constant rate__).

### **Key Differences between Pearson correlation and Spearman correlation:**

- Pearson correlation is suitable for assessing linear relationships in normally distributed data, while Spearman correlation is suitable for monotonic relationships and is less affected by outliers.


- Pearson correlation uses the actual values of the variables, whereas Spearman correlation uses the rank order of the data.


- Both coefficients range from -1 to 1, but the interpretation of their values is different: Pearson assesses linear relationships, while Spearman assesses monotonic relationships.



In summary, Pearson is more appropriate for linear relationships in normally distributed data, while Spearman is versatile and suitable for a broader range of data types and relationships.

### Correlation and Causation

The phrase "correlation does not imply causation" means that just because
two variables are associated with each other, it does not necessarily mean that
one causes the other. In other words, a correlation between two variables
does not necessarily imply that one variable is the reason for the other
variable's behaviour.



Suppose there is a positive correlation between the number of firefighters
present at a fire and the amount of damage caused by the fire. One might be
tempted to conclude that the presence of firefighters causes more damage.
However, this correlation could be explained by a third variable - the severity
of the fire. More severe fires might require more firefighters to be present, and
also cause more damage.



Thus, while correlations can provide valuable insights into how different
variables are related, they cannot be used to establish causality. Establishing
causality often requires additional evidence such as experiments, randomized
controlled trials, or well-designed observational studies.

## Quantiles and Percentiles

Quantiles are statistical measures used to divide a set of numerical data into
equal-sized groups, with each group containing an equal number of observations.


Quantiles are important measures of variability and can be used to: understand
distribution of data, summarize and compare different datasets. They can also be
used to identify outliers.


There are several types of quantiles used in statistical analysis, including:

- __Quartiles:__ Divide the data into four equal parts, Q1 (25th percentile), Q2
(50th percentile or median), and Q3 (75th percentile).


- __Deciles:__ Divide the data into ten equal parts, D1 (10th percentile), D2
(20th percentile), ..., D9 (90th percentile).


- __Percentiles:__ Divide the data into 100 equal parts, P1 (1st percentile), P2
(2nd percentile), ..., P99 (99th percentile).


- __Quintiles:__ Divides the data into 5 equal parts


### NOTE : Things to remember while calculating these measures:

1. __Data should be sorted from low to high__


2. You are basically finding the location of an observation


3. They are not actual values in the data


4. All other tiles can be easily derived from Percentiles

### Percentile

A percentile is a statistical measure that represents the percentage of observations in a
dataset that fall below a particular value. 

For example, the 75th percentile is the value below

Formula to calculate the percentile value:

$$PL = \frac{p}{100} *(N+1)$$

### Difference between percentile and percentage:


**Scenario**: Consider a class of 20 students who recently took a math test. You want to understand how each student performed in relation to the rest of the class.

**Percentage**:

- Percentage represents a portion of a whole, typically expressed as a fraction of 100.
- It is used to measure the relative quantity of something out of 100%.
- In this context, percentage is used to determine how many students scored above a certain threshold.

**Example of Percentage**:
   - If 15 out of 20 students scored above 75 marks, you can calculate the percentage of students who scored above 75 as follows:
   
     Percentage = (Number of Students Above 75 / Total Number of Students) * 100%
     
     Percentage = (15 / 20) * 100% = 75%
     
   - This means that 75% of the students in the class scored above 75 marks on the math test.

**Percentile**:

- Percentile represents a specific data point's position within a dataset, indicating the percentage of data points that are equal to or below it.
- It is used to understand a data point's relative position within a dataset, particularly in comparison to others.
- In this context, percentile is used to determine how well a student's score ranks within the class.

**Example of Percentile**:
   - If you want to find the 80th percentile of the class scores, you arrange all the scores in ascending order and find the score below which 80% of the scores fall.
   
     Arrange scores: [60, 65, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 94, 95, 96, 97, 98, 99, 100, 100]
   
     The 80th percentile score is 94. This means that 80% of the students scored 94 marks or lower.

In summary:

- **Percentage** is used to express a portion out of 100 and measures relative quantities or proportions.
- **Percentile** is used to find a specific data point's relative position within a dataset, indicating the percentage of data points that are equal to or below it.

In our example, percentage tells us how many students scored above a certain threshold, while percentile tells us where a specific student's score ranks relative to the entire class.

## 5 number Summary

The five-number summary is a descriptive statistic that provides a summary of a
dataset. It consists of five values that divide the dataset into four equal parts, also
known as quartiles. The five-number summary includes the following values:

1. __Minimum value:__ The smallest value in the dataset.


2. __First quartile (Q1):__ The value that separates the lowest 25% of the data from
the rest of the dataset.


3. __Median (Q2):__ The value that separates the lowest 50% from the highest 50%
of the data.


4. __Third quartile (Q3):__ The value that separates the lowest 75% of the data from
the highest 25% of the data.


5. __Maximum value:__ The largest value in the dataset.


The five-number summary is often represented visually using a box plot, which
displays the range of the dataset, the median, and the quartiles.


The five-number summary is a useful way to quickly summarize the central
tendency, variability, and distribution of a dataset.

### Removing Outliers:

Anything outside Lower and Higher Fence is known as Outlier.

$$OUTLIER \longleftarrow [Lower \;Fence \;and \;Higher\;Fence]\longrightarrow OUTLIER $$

##### Interquartile Range:
The interquartile range (IQR) is a measure of variability that is based on the five-number
summary of a dataset. Specifically, the IQR is defined as the difference between the third quartile (Q3) and the first quartile (Q1) of a dataset.

<br></br>
$\text{Lower Fence} = \text{Q1 - (1.5 * IQR)}$

$\text{Higher Fence} = \text{Q3 + (1.5 * IQR)}$

$\text{where IQR is Inter Quartile Range}=\;Q3\;value\;-\;Q1\;value$

$\text{Q1 is 25th percentile positon's value, Q3 is 75th percentile positon's value}$

#### why 1.5 is used in terms of IQR?
The factor of 1.5 is commonly used in the context of the Interquartile Range (IQR) to define outliers and is part of the Tukey's Fences or Tukey's Rule for identifying outliers. 


The factor of 1.5 is somewhat arbitrary but is commonly used because it provides a balance between identifying outliers while not being too sensitive to slight variations in the data. It's a rule of thumb that is considered robust for many datasets.
