### Sampling Distribution

Sampling distribution is a probability distribution that describes the statistical properties of a
sample statistic (such as the sample mean or sample variance) computed from multiple
independent samples of the same size from a population.

___NOTE : size of each sample should be equal___

### Why Sampling Distribution is important?

Sampling distribution is important in statistics and machine learning because it allows us to
estimate the variability of a sample statistic, which is useful for __making inferences about the
population.__

By analysing the properties of the sampling distribution, we can compute
confidence intervals, perform hypothesis tests, and make predictions about the population
based on the sample data.

__Sampling Distribution of mean : When we are referring to a distribution form by samples of means of same sizes, we use the term a sampling distribution.__

$\text{sampling distribution of sample means}\;=\bar X_1, \bar X_2, \bar X_3, \bar X_4......\bar X_{99},\bar X_{100}$

$\text{sampling distribution of sample std dev}\;=\bar \sigma_1,\bar \sigma_2,\bar \sigma_3,\bar \sigma_4......\bar \sigma_{99},\bar \sigma_{100}$

## Central Limit Theorem

The Central Limit Theorem (CLT) states that the distribution of the __sample means of a large
number of independent and identically distributed random variables will approach a normal
distribution,__ regardless of the underlying distribution of the variables.

The conditions required for the CLT to hold are:
1. The sample size is large enough, typically greater than or equal to 30.


2. The sample is drawn from a finite population or an infinite population with a finite
variance.


3. The random variables in the sample are independent and not baised.

## <span class="mark">Standard Error:</span>

- The Standard Error (SE) is a measure of the variability or precision of a sample statistic, such as the sample mean or sample proportion, when compared to the true population parameter. 


- __The Standard Error is a measure of how much the sample mean (or other sample statistics) varies from sample to sample.__
In the context of the Central Limit Theorem, the Standard Error tells us how much we can expect the sample means to deviate from the true population mean on average.



- As the sample size $(n)$ increases, the Standard Error decreases. This means that larger samples provide more precise estimates of the population parameter (e.g., the population mean).


- The Standard Error is directly related to the standard deviation of the population and inversely related to the square root of the sample size:

  $$SE = \frac{S_{avg}}{\sqrt{n}}$$

  Where:
  
  
  - $SE$ is the Standard Error of the sample mean.
  
  
  - $S$ is the __average standard deviation of the sample means__.
  
  
  - $(n)$ is the sample size.



- When the sample size is large (as implied by the CLT), the Standard Error becomes small, indicating that the sample mean is likely to be very close to the population mean.



- For example, when we calculate the average score of a sample of students and report it as the sample mean, we use the Standard Error to understand how much the sample mean might vary if we were to take many different random samples from the same population.

### Difference between Standard Error and Standard error of sample means:

The Standard Error (SE) and the Standard Error of the Sample Mean (SE of the Sample Mean) are related concepts but have different formulas and purposes. Here's an explanation of each, their formulas, and when they are used:

**1. Standard Error (SE):**

- **Formula**: The formula for the Standard Error (SE) depends on what you are calculating the SE for. In general, it is a measure of the variability of a sample statistic (e.g., sample mean or sample proportion) as an estimate of the population parameter.

  
  
  - For the sample mean $(SE_{\bar{x}})$:
    $SE_{\bar{x}} = \frac{s}{\sqrt{n}}$
    
  - For the sample proportion $(SE_p)$ when estimating a population proportion:
    $SE_p = \sqrt{\frac{p(1-p)}{n}}$
    
- Where:   
 
     - s is the sample standard deviation.    
    
     - n is the sample size.
        
     - p is the estimated population proportion.





**2. Standard Error of the Sample Mean (SE of the Sample Mean):**

- **Formula**: The Standard Error of the Sample Mean specifically refers to the SE when estimating the population mean $(\mu)$ using the sample mean $(\bar{x})$. It's calculated as:

  $SE_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$

  - Where:
    
    - $\sigma$ is the population standard deviation.
    
    - $(n)$ is the sample size.




**Differences and Usage**:

- The primary difference is that the SE of the Sample Mean uses the population standard deviation $(\sigma)$, while the general Standard Error uses the sample standard deviation $(s)$.
  
- The SE of the Sample Mean is typically used when you know the population standard deviation or when you're working with a theoretical or idealized scenario.

- The general Standard Error is used when you don't know the population standard deviation and you're estimating it using the sample standard deviation. This is the more common scenario in practical data analysis.

In summary, both the Standard Error and the SE of the Sample Mean are used to quantify the variability or precision of sample statistics when estimating population parameters. The choice between them depends on whether you have information about the population standard deviation (SE of the Sample Mean) or not (general Standard Error).

**Relation Between SE and σ**:
   
   
   - The standard error (SE) is related to the population standard deviation (σ) and the sample size (n) by the following formula:
   
     $$SE = \frac{σ}{√n}\;or\;\frac{\sigma^2}{n}$$

   
   - As the sample size (n) increases, the standard error (SE) decreases. In other words, larger samples result in sample means that are closer to the population mean.



The relationship between the population standard deviation (σ) and the standard error (SE) is important because it quantifies the precision of our sample means. A larger sample size (n) reduces the standard error, making our sample means a more accurate estimate of the population mean (μ). This is why larger sample sizes are often preferred when conducting statistical analyses.

### Using confidence and probability with mean of sample means:

### Case Study - What is the average income of Indians
#### Step-by-step process:


1. Collect multiple random samples of salaries from a representative group
of Indians. Each sample should be large enough (usually, n > 30) to ensure
the CLT holds. Make sure the samples are representative and unbiased to
avoid skewed results.


2. Calculate the sample mean (average salary) and sample standard
deviation for each sample.


3. Calculate the average of the sample means. This value will be your best
estimate of the population mean (average salary of all Indians).


4. Calculate the standard error of the sample means, which is the standard
deviation of the sample means divided by the square root of the number
of samples.


5. Calculate the confidence interval around the average of the sample means
to get a range within which the true population mean likely falls. For a
95% confidence interval:


__lower_limit = average_sample_means - 1.96 * standard_error__


__upper_limit = average_sample_means + 1.96 * standard_error__


___1.96 is the reliability factor___

6. Report the estimated average salary and the confidence interval.

$$\text{margin of error = reliability factor * Standard Error}$$

````python
import numpy as np

# Set the parameters
population_size = 100000
sample_size = 50
num_samples = 100

# Generate a random representative sample of salaries (in thousands)
# You should replace this with actual collected salary data
np.random.seed(42)  # Setting a seed for reproducibility
population_salaries = np.random.lognormal(mean=4.5, sigma=0.8, size=population_size)
print("population mean =",np.mean(population_salaries),"\n")
print("population std dev =",(np.std(population_salaries)),"\n")

# Generate multiple samples and calculate the sample means and standard deviations
sample_means = []
sample_std_devs = []

for _ in range(num_samples):
    sample_salaries = np.random.choice(population_salaries, size=sample_size)
    sample_means.append(np.mean(sample_salaries))
    sample_std_devs.append(np.std(sample_salaries))

# Calculate the average of the sample means and the standard error
average_sample_means = np.mean(sample_means)
print("mean of sample means =",average_sample_means)

standard_error = np.std(sample_means) / np.sqrt(num_samples)
print("standard error =",np.mean(standard_error),"\n")


# Calculate the 95% confidence interval
margin_of_error = 1.96 * standard_error
lower_limit = average_sample_means - margin_of_error
upper_limit = average_sample_means + margin_of_error

# Report the results
print(f"Estimated average salary (in thousands): {average_sample_means:.2f}")
print(f"95% confidence interval (in thousands): ({lower_limit:.2f}, {upper_limit:.2f})")
````

### <span class="mark">Explanation for confusion between pop std dev and std dev of sample of means:</span>

#### In the code you provided, you are correctly calculating the standard error (SE) using the formula:

```
standard_error = np.std(sample_means) / np.sqrt(num_samples)
```

This formula is appropriate for calculating the standard error of the sample mean, and it correctly divides the sample standard deviation by the square root of the sample size $(\sqrt{N})$.

However, the confusion might arise from comparing the population standard deviation $(\sigma)$ to the standard error of the sample mean $(SE(\bar{x}))$. These two values serve different purposes:

1. **Population Standard Deviation $(\sigma)$:**
   - This measures the dispersion or variability of the entire population.
   - It tells you how much individual data points in the entire population deviate from the population mean.
   - In your code, you correctly calculate the population standard deviation using `np.std(population_salaries)`.



2. **Standard Error of the Sample Mean $(SE(\bar{x}))$:**
   - This measures the precision of the sample mean as an estimate of the population mean.
   - It tells you how much the sample means would vary if you took multiple random samples of the same size from the same population.
   - It is calculated by dividing the sample standard deviation by the square root of the sample size, as you did in your code.

The key distinction is that the population standard deviation $(\sigma)$ pertains to the entire population, while the standard error $(SE(\bar{x}))$ pertains to the precision of the sample mean as an estimate of the population mean.

It is normal for the population standard deviation $(\sigma)$ to be different from the standard error $(SE(\bar{x}))$ of the sample mean, especially when dealing with finite sample sizes. The standard error is affected by both the population standard deviation and the sample size, so it provides a measure of how much the sample means are likely to vary due to random sampling, not the spread of the entire population.

In an idealized scenario with an infinite sample size (i.e., as the sample size approaches the population size), the standard error $(SE(\bar{x}))$ would approach $\sigma / \sqrt{N}$, but in practice, with finite sample sizes, there will be some variation due to sampling.

### Population Vs Sample

__Population:__ A population is the entire group or set of individuals, objects, or events that a
researcher wants to study or draw conclusions about. It can be people, animals, plants, or
even inanimate objects, depending on the context of the study. The population usually
represents the complete set of possible data points or observations.


__Sample:__ A sample is a subset of the population that is selected for study. It is a smaller group
that is intended to be representative of the larger population. Researchers collect data from
the sample and use it to make inferences about the population as a whole. Since it is often
impractical or impossible to collect data from every member of a population, samples are used
as an efficient and cost-effective way to gather information.

### Parameter Vs Statistic
__Parameter:__ A parameter is a numerical value that describes a characteristic of a population.
Parameters are usually denoted using Greek letters, such as μ (mu) for the population mean or
σ (sigma) for the population standard deviation. Since it is often difficult or impossible to
obtain data from an entire population, parameters are usually unknown and must be
estimated based on available sample data.


__Estimate / Statistic:__ A statistic is a numerical value that describes a characteristic of a sample, which is a
subset of the population. By using statistics calculated from a representative sample,
researchers can make inferences about the unknown respective parameter of the population.
Common statistics include the sample mean (denoted by , pronounced "x-bar"), the sample
median, and the sample standard deviation (denoted by s).

An estimate is the output that you get from the estimator (when you apply the formula).

There are two types of estimates, point estimates and confidence interval estimates.
- **Point Estimates :** The value of any statistic that estimates the value of a parameter. 
                    : A single value. eg: 1, 2, 5.5, 0.32
- **Confidence Intervals :** An Interval. eg: (1,5), (12,33), (-0.71,0.11)

Confidence intervals are much more precise than point estimates. That is why they are preferred when making inferences

#### Inferential Statistics
Inferential statistics is a branch of statistics that focuses on making predictions, estimations, or
generalizations about a larger population based on a sample of data taken from that
population. It involves the use of probability theory to make inferences and draw conclusions
about the characteristics of a population by analysing a smaller subset or sample.

The key idea behind inferential statistics is that it is often impractical or impossible to collect
data from every member of a population, so instead, we use a representative sample to make
inferences about the entire group. Inferential statistical techniques include hypothesis testing,
confidence intervals, and regression analysis, among others.

These methods help researchers answer questions like:

a. Is there a significant difference between two groups?

b. Can we predict the outcome of a variable based on the values of other variables?

c. What is the relationship between two or more variables?

Inferential statistics are widely used in various fields, such as economics, social sciences,
medicine, and natural sciences, to make informed decisions and guide policy based on limited
data.

### Point Estimate

A point estimate is a single value, __calculated from a sample,__ that serves as the best guess or
approximation for an unknown population parameter, such as the mean or standard
deviation. Point estimates are often used in statistics when we want to __make inferences about
a population based on a sample.__

eg : sample mean $(\bar x)$, sample std dev $\bar \sigma$

## Confidence Interval

__Confidence interval__, in simple words, is a range of values within which we expect a particular
population parameter, like a mean, to fall. It's a way to express the uncertainty around an
estimate obtained from a sample of data.


__Confidence level__, usually expressed as a percentage like 95%, indicates how sure we are that
the true value lies within the interval.

#### Margin of Error (ME)

Getting a smaller margin of error means that the confidence interval would be narrower as we want a better prediction, it is in our interest to have the narrowest possible confidence interval.

The best part is that we can control the margin of error.
 
 The true population mean falls in the interval defined by the 
 
 $$sample\; mean \;(\bar \chi)\;\pm \text{the margin of error}$$
 
 where $$\text{margin of error = reliability factor}\;*\; \text{Standard Error}$$
 
 $$\text{reliability factor} =\; \text{t-score} \; \longrightarrow \left\lceil t_{n-1},\frac{_\alpha}{_2}\right\rceil\;\;or\;\;\text{z-score}   \longrightarrow \left\lfloor z_\left(\frac{\alpha}{2}\right)\right\rfloor$$
 
 $$Standard\;Error\longrightarrow\frac{standard\;deviation}{\sqrt {sample\;size}}\longrightarrow\frac{\sigma}{\sqrt n}$$
 
<br></br> 
A higher level of confidence$(1-\alpha)$ increases the statistic, a higher statistic means a higher margin of error. This leads to a wider confidence interval.
<br></br>
A lower standard deviation$(\sigma)$ means that the data set is more concentrated around the mean. So we have a better chance to get it right.
<br></br>
Higher sample sizes$(n)$ will decrease the margin of error. The more observations you have in your sample, the more certain you are in the prediction.

        - A higher statistic increases the Margin of Error (ME). 
        - A higher standard deviation increases the  ME. 
        - A higher sample size decreases the ME.

## Z-Procedure Confidence Interval (Sigma Known)

### Assumptions


1. __Random sampling:__ The data must be collected using a random sampling method to
ensure that the sample is representative of the population. This helps to minimize biases
and ensures that the results can be generalized to the entire population.


2. __Known population standard deviation:__ The population standard deviation (σ) must be
known or accurately estimated. In practice, the population standard deviation is often
unknown, and the sample standard deviation (s) is used as an estimate. However, if the
sample size is large enough, the sample standard deviation can provide a reasonably
accurate approximation.


3. __Normal distribution or large sample size:__ The Z-procedure assumes that the underlying
population is normally distributed. However, if the population distribution is not normal,
the Central Limit Theorem can be applied when the sample size is large (usually, sample
size n ≥ 30 is considered large enough). According to the Central Limit Theorem, the
sampling distribution of the sample mean will approach a normal distribution as the
sample size increases, regardless of the shape of the population distribution.


 We build the confidence interval around the point estimate.

### Interpreting Confidence Interval

A confidence interval is a range of values within which a population parameter, such as the
population mean, is estimated to lie with a certain level of confidence. The confidence interval
provides an indication of the precision and uncertainty associated with the estimate. To
interpret the confidence interval values, consider the following points:

1. __Confidence level:__ The confidence level (commonly set at 90%, 95%, or 99%) represents
the probability that the confidence interval will contain the true population parameter if
the sampling and estimation process were repeated multiple times. For example, a 95%
confidence interval means that if you were to draw 100 different samples from the
population and calculate the confidence interval for each, approximately 95 of those
intervals would contain the true population parameter.


2. __Interval range:__ The width of the confidence interval gives an indication of the precision of
the estimate. A narrower confidence interval suggests a more precise estimate of the
population parameter, while a wider interval indicates greater uncertainty. The width of
the interval depends on the sample size, variability in the data, and the desired level of
confidence.


3. __Interpretation:__ To interpret the confidence interval values, you can say that you are "X%
confident that the true population parameter lies within the range (lower limit, upper
limit)." Keep in mind that this statement is about the interval, not the specific point
estimate, and it refers to the confidence level you chose when constructing the interval.

### calculating sample size for estimating a population mean with a confidence interval :


The sample size formula for calculating the required sample size for a confidence interval depends on several factors, including the __desired level of confidence, the margin of error you can tolerate, and the population standard deviation (if known).__ The general formula for calculating the sample size for estimating a population mean with a confidence interval is:

$$n = \frac{{(Z \cdot \sigma)^2}}{{E^2}}$$

Where:
- $n$ = required sample size
- $Z$ = Z-score corresponding to the desired level of confidence (e.g., for a 95% confidence level, $Z$ might be 1.96, which corresponds to the 97.5th percentile of the standard normal distribution)
- $\sigma$ = population standard deviation (if known)
- $E$ = desired margin of error (the maximum acceptable difference between the sample estimate and the true population parameter)

If you don't know the population standard deviation $\sigma$, you can use the sample standard deviation $s$ as an estimate in its place. In that case, you would use the following formula:

$$n = \frac{{(Z \cdot s)^2}}{{E^2}}$$

To calculate the sample size, you need to decide on the level of confidence you want (usually expressed as a percentage, such as 95% or 99%), determine the margin of error that is acceptable for your study, and, if you know it, use the population standard deviation or an estimate based on a preliminary sample.

Here are the steps to calculate the sample size:

1. Choose your desired level of confidence (e.g., 95%, 99%, etc.).
2. Determine the margin of error (E) that you can tolerate.
3. If you know the population standard deviation $(\sigma)$, use it. If not, use a preliminary estimate based on a sample $s$.
4. Look up the appropriate Z-score for your chosen level of confidence (you can find this in standard normal distribution tables or use a calculator).
5. Plug these values into the formula to calculate the required sample size (n).

Keep in mind that increasing the level of confidence or decreasing the margin of error will require a larger sample size. Conversely, reducing the level of confidence or increasing the margin of error will lead to a smaller required sample size.

## t-procedure - Confidence Interval (Sigma not known)

___Mostly used as in real life we don't get std dev of population (sigma)___

### Assumptions:

1. __Random sampling:__ The data must be collected using a random sampling method to
ensure that the sample is representative of the population. This helps to minimize biases
and ensures that the results can be generalized to the entire population.


2. __Sample standard deviation:__ The population standard deviation (σ) is unknown, and the
sample standard deviation (s) is used as an estimate. The t-distribution is specifically
designed to account for the additional uncertainty introduced by using the sample
standard deviation instead of the population standard deviation.


3. __Approximately normal distribution:__ sample size should be > 30 if it's not normal. The t-procedure assumes that the underlying
population is approximately normally distributed, or the sample size is large enough for
the Central Limit Theorem to apply. If the population distribution is heavily skewed or has
extreme outliers, the t-procedure may not be accurate, and non-parametric methods
should be considered.


4. __Independent observations:__ The observations in the sample should be independent of
each other. In other words, the value of one observation should not influence the value of
another observation. This is particularly important when working with time series data or
data with inherent dependencies.

$$\text{Degree of Freedom}\;=\;(n-1)$$

$$\text{Where n is sample size}$$

Here we don't have population std dev. so we will use sample std dev



__Formula:__

$$t = \frac{\bar x - \mu}{\frac{s_{avg}}{\sqrt n}}$$





- t is the t-statistic.


- x is the sample mean.


- μ is the hypothesized population mean.


- __s is the avg of all sample std deviation.__


- n is the sample size.

### Student's T Distribution


- Visually, the students-t distribution looks much like a normal distribution, __but generally has fatter tails.__


- Fatter tales, allows for a higher dispersion of variables and there is more uncertainty. This means the probability of values being far away from the mean is bigger. 


- For big enough samples, the Student’s T distribution coincides with the Normal distribution. In the same way that the z- statistic is related to the standard normal distribution, the T- statistic is related to the students t distribution.


- The Student’s T distribution is used predominantly for creating confidence intervals and testing hypotheses with normally distributed populations when the sample sizes are small. It is particularly useful when we don’t have enough information or it is too costly to obtain it.

All else equal, the Student’s T distribution has fatter tails than the Normal distribution and a lower peak This is to reflect the higher level of uncertainty, caused by the small sample size.


For Sample of $n$ **the degree of freedom =** $(n-1)$
- eg: for sample of 20 observations, degree of freedom is 19.

>**Note :** After 30 degrees of freedom, the T-Statistic table becomes almost the same as the Z statistic.As the degrees of freedom depend on the sample, in essence, the bigger the sample, the closer we get to the actual numbers. A common rule of thumb is that for a sample containing more than 50 observations, we use the Z table instead of the t-table.

## When to use z-test or t-test :

The key points to consider when deciding whether to use a z-test or a t-test:

__1. Sample Size:__
   - Z-Test: Typically used when you have a large sample size (usually n ≥ 30). This is because the z-test assumes that the sampling distribution of the sample mean is approximately normally distributed, which holds true for large sample sizes due to the central limit theorem.
   - T-Test: Generally used when you have a small sample size (typically n < 30). The t-test is more robust to deviations from normality in small samples.



__2. Population Standard Deviation:__
   - Z-Test: Appropriate when you know the population standard deviation (σ) or have a large sample size and can assume the population standard deviation based on your sample.
   - T-Test: Used when you do not know the population standard deviation and must estimate it from the sample (s). The t-test uses the sample standard deviation to make inferences.



__3. Type of Data:__
   - Z-Test: Suitable for situations where the data is normally distributed or when the sample size is large enough for the central limit theorem to apply.
   - T-Test: More flexible and can be used for data that may not be perfectly normally distributed, especially with smaller sample sizes.



__4. One-Sample vs. Two-Sample:__
   - Z-Test: Typically used for one-sample tests, where you are comparing the sample mean to a known population mean.
   - T-Test: Commonly used for two-sample tests, comparing the means of two independent groups or paired samples.



__5. Independence of Samples:__
   - Z-Test: Appropriate when dealing with independent samples or situations where the observations are not paired.
   - T-Test: Used when comparing paired or dependent samples, such as before-and-after measurements or matched pairs.



__6. Assumptions:__
   - Z-Test: Assumes that the population follows a normal distribution and that you know the population standard deviation.
   - T-Test: Assumes that the population follows a normal distribution (or the sample size is large enough for the central limit theorem to apply) and that the data within each group is independent and has equal variances (homoscedasticity).



__7. Degrees of Freedom:__
   - Z-Test: Does not involve degrees of freedom in its formula.
   - T-Test: The degrees of freedom vary depending on the type of t-test (e.g., independent samples vs. paired samples) and affect the shape of the t-distribution.
<br></br>
In summary, if you have a large sample size and know the population standard deviation, you may opt for a z-test. For smaller sample sizes or when you don't know the population standard deviation, the t-test is generally more appropriate. Additionally, consider the distribution of your data and whether the assumptions of each test are met.

### when to use what?

It seems you might be referring to two terms: "t-procedure" and "z-procedure." These terms are often used in the context of hypothesis testing and confidence intervals when dealing with sample means. Let me explain each and provide an example of when to use them:

1. **t-Procedure (t-Test):**
   - The t-procedure, also known as a t-test, is used when you want to make inferences about a population mean based on a sample mean, especially when you don't know the population standard deviation.
   - It's used when working with small samples (typically n < 30) or when the population standard deviation is unknown.
   - There are different types of t-tests, such as the one-sample t-test, independent samples t-test (for comparing means of two independent groups), and paired samples t-test (for comparing means of paired data points).
   
   **Example:** You want to know if the average height of a sample of 20 students is significantly different from the population average height when the population standard deviation is unknown.
$$t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}$$

Where:
- t is the t-statistic.
- $\bar{x}$ is the sample mean.
- $mu$is the population mean under the null hypothesis.
- s is the sample standard deviation.
- n is the sample size.


2. **z-Procedure (z-Test):**
   - The z-procedure, also known as a z-test, is used when you want to make inferences about a population mean based on a sample mean, and you know the population standard deviation.
   - It's typically used when dealing with larger samples (n ≥ 30) and you have information about the population standard deviation.
   
   **Example:** You want to know if the average test scores of a sample of 100 students are significantly different from the population average test score when the population standard deviation is known.


$$z= \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}$$

Where:
- z is the z-statistic.
- $\bar{x}$ is the sample mean.
- $\mu$ is the population mean under the null hypothesis.
- $\sigma$ is the population standard deviation.
- n is the sample size.

In summary, you use the t-procedure (t-test) when dealing with small samples or when the population standard deviation is unknown, and you use the z-procedure (z-test) when working with larger samples and you know the population standard deviation. The choice between them depends on the characteristics of your data and the information available about the population.

### z-table : 

#### +ve z-table : 

````python
import pandas as pd
import numpy as np
import scipy.stats as st

# Define the range of z-scores to be included in the table
z_range_p_ve = np.arange(0.0, 3.5, 0.1)
col = np.arange(0.00, 1.0, 0.1)/10
# Create a pandas dataframe to store the z-score table
z_table = pd.DataFrame(index=z_range_p_ve, columns=col)

# Fill the dataframe with the corresponding probability values
for i in z_range_p_ve:
    for j in col:
        z_table.loc[i,j] = round(st.norm.cdf(j+i), 4)
print("**+ve z- score table**")
# Print the z-score table
z_table
````

#### -ve z-table : 

````python

pd.options.display.float_format = '{:f}'.format

# Define the range of z-scores to be included in the table
z_range_n_ve = np.arange(-3.4, 0.1, 0.1)
col = np.arange(0.00, 1.0, 0.1)/10
# Create a pandas dataframe to store the z-score table
z_table = pd.DataFrame(index=z_range_n_ve, columns=col)

# Fill the dataframe with the corresponding probability values
for i in z_range_n_ve:
    for j in col:
        z_table.loc[i,j] = round(st.norm.cdf(i-j), 4)
print("-ve z-score table")
# Print the z-score table
z_table
````
