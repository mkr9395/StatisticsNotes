
#### 4. Poisson Distribution
<br></br>
The Poisson Distribution deals with the frequency with which an event occurs within a specific interval. Instead of the probability of an event, the Poisson Distribution requires knowing how often it occurs for a specific period of time or distance.
<br></br>
>$$It\; is \; denoted\; by\;\; Po(\lambda)$$
$$Y \sim Po(4)$$

We read above statement as variable y follows a poisson distribution with Lambda equal to four.

>$$Formula\;\;P(Y) = \frac{\lambda ^ y * e ^ {-\lambda}}{y!}$$


$e \rightarrow Euler's\;number\;(Napier's\;Constant), e\approx 2.72\;\;and\;\; e^{-\lambda}=\frac{1}{e^\lambda}$

<br></br>
eg:imagine you created an online course on probability, usually your students asked you around four questions per day, but yesterday, they asked seven. Surprised by this sudden spike in interest from your students, you wonder how likely it was that they would ask exactly seven questions 
i.e $P(y=7)=?$
$$average\;no.\;of\;questions\;here\;is\; \lambda \rightarrow 4$$
$$time\;interval\;is\;entire\;1\;day$$
$$the\;singular\;instance\;you\;are\;interested\;in\;is\;y \rightarrow 7$$
$$So,\; P(7) = \frac{4^7\;*\;e^{-4}}{7!}\;\;\approx 0.06 $$

So there is 6% chance of receiving exactly 7 questions.

<br></br>
**Expected Value for Poisson Distribution**

$$E(Y) = \lambda$$

**Variance for Poisson Distribution**
$$\sigma^2 = \lambda$$

$$So \;\; mean\;(\mu) = Variance\;(\sigma^2) = \lambda$$
<br><br>
    
    Q. Why does the graph of the Poisson Distribution start from 0 on the X-axis?

Because we are measuring the frequency of occurrence over a given time interval or distance.

#### Exponential Distribution
$$Denoted\;by\; Exp(\lambda)$$
$$X \sim Exp (\frac{1}{2})$$
*variable X follows an exponential distribution lambda.*

<br></br>
Variables which most closely follow an exponential distribution are ones with a probability that initially decreases before eventually plateauing. 

Eg: One such example is the number of views for a YouTube Vlog.There is a great interest upon release, so it starts off with many views in the first day or two. After most subscribers have had the chance to see the video, the view counter slows down. Even though the aggregate amount of views keeps increasing, the number of new ones diminishes daily. As time goes on, the video either becomes outdated or the author produces new content, so viewership focus shifts away.

Therefore, it is more likely for a random viewing to have occurred close to the video's initial release

<br></br>
Graphically, the **PDF** of such a function would start off very high and sharply decrease within the first few time frames, the curve somewhat resembles a boomerang with each handle lining up with the X and Y axis.


In a weird way, the **CDF** would also resemble a boomerang. However, this one has shifted 90 degrees to the right. As you know, the cumulative distribution eventually approaches one, so that would be the value where it plateau.

<br></br>

To define an exponential distribution, we require a rate parameter denoted by  lambda $(\lambda)$. This parameter determines how fast the CDV curve reaches the point of plateauing and how spread out the graph is.

* $Expected\;Value\;of\;Exponential\;Distribution\;is\;\; E(Y) = \frac{1}{\lambda}$ .
* $Var(Y) = \frac{1}{\lambda^2}$.


However, unlike the normal $(N)$ or Chi-squared distributions $(\chi^2)$, we do not have a table of known variables for it.
That is why sometimes we prefer to transform it.

Generally, we can take the natural logarithm of every set of an exponential distribution  and get a normal distribution in statistics, we can use this new transform data to run linear regressions. This is one of the most common transformations I've had to perform.

$$Y \sim\;Exp(\lambda),\;\;\; X = ln(y) \;\;\Rightarrow X \sim N (\mu, \sigma^2)$$

        graphs of the Exponential Distribution?
        The CDF plateaus near 1 mark and the PDF plateaus around the 0 mark.

#### Logistic Distribution

$$Denoted\;by\;\;Logistic(\mu, S)$$
$$where\;S\;is\;Scale\;Parameter\;and\;\mu\;is\;Location$$

$$Y\;\sim \;\;Logistic(6, 3)$$
we read the statement below as variable y follows a logistic distribution with location six and a scale of three.

We often encounter logistic distributions when trying to determine how continuous variable inputs can affect the probability of a binary outcome. This approach is commonly found in forecasting competitive sports events where there exist only two clear outcomes victory or defeat.

eg: For instance, we can analyze whether the average speed of a tennis player serve plays a crucial role in the outcome of the match. Expectation dictates that sending the ball with higher velocity leaves opponents with a shorter period to respond. This usually results in a better hit, which could lead to a point for the server.
To reach the highest speeds, tennis players often give up some control over the shot, so our less accurate.

Therefore, we cannot assume that there is a linear relationship between point conversion and serves beats.
Theory suggests there exists some optimal speed which enables the serve to still be accurate enough. Then most of the shots we convert into points will likely have similar velocities.

As tennis players go further away from the optimal speed, their shots either become too slow and easy to handle or to an accurate. This suggests that the graph of the **PDF** of the logistic distribution would look similarly to the **normal distribution.**


Actually, the graph of the logistic distribution is defined by two key features, its mean$(\mu)$ and its scale parameter$(S)$. The former $(\mu)$ dictator at the center of the graph, whilst the latter $(S)$ shows how spread out the graph is going to be.

Going back to the tennis example, the mean would represent the optimal speed whilst the scale would dictate how lenient we can be with the hit.To elaborate, some tennis players can hit a great serve further away from the optimal speed than others. For instance, Serena Williams can hit fantastic serves even if the ball moves faster or slower than
it optimally should. Therefore, she is going to have a more spread out PDF than some of her opponents.

<br></br>
Now let's discuss the **cumulative distribution function (CDF).**

It should be a curve that starts off slow, then picks up rather quickly before plateauing around the one mark.
That is because once we reach values near the mean, the probability of converting the point drastically goes up.
Once again, the scale would dictate the shape of the graph. In this case, the smaller the scale, the later the graft starts to pick up, but the quicker it reaches values close to one.
- It follows an S-shape and plateaus around the value 1.
- The probability drastically starts to increase once we reach values close to mean
- The steeper the curve, the faster it reaches value close to absolute certainity (1).

<br></br>
* $ Expected\;Value\;of\;Logistic\;Distribution\;is\;\; E(Y) = \mu$
* $Var(Y) = \frac{S^2 * \pi^2}{3} \;\;where\; S^2\;is\;square\;of\;scale$

### Chebyshev's Inequality

**Imperical Formula**

$If\;X\approx\;Gaussian\;Distribution$

$Probability\;(\mu\;-\;\sigma\;\leqslant\;\chi\;\leqslant\mu\;+\;\sigma)\;\;\longrightarrow\;68\%$

$Probability\;(\mu\;-\;2\sigma\;\leqslant\;\chi\;\leqslant\mu\;+\;2\sigma)\;\;\longrightarrow\;95\%$

$Probability\;(mu\;-\;3\sigma\;\leqslant\;\chi\;\leqslant\mu\;+\;3\sigma)\;\;\longrightarrow\;99.7\%$
<br></br>

- $But\;If\;Y \neq Gaussian\;Distribution$

$$Probability\;(\mu\;-\;k\sigma\;\leqslant\;y\;\leqslant\mu\;+\;k\sigma)\;\geqslant\;(1-\frac{1}{K^2})$$

$where\;"k"\;may\;be\;standard\;deviation\;and\;value\;of\;k\;\geq\;2$
<br></br>

- so for 2 standard deviation i.e. k = 2 in Chebyshev equation: 

$$Probability\;(\mu\;-\;2\sigma\;\leqslant\;y\;\leqslant\mu\;+\;2\sigma)\;\geqslant\;(1-\frac{1}{K^2})\longrightarrow(1-\frac{1}{4})\longrightarrow(\frac{3}{4})\longrightarrow\;75\%$$

- so for 3 standard deviation i.e. k = 3 in Chebyshev equation: 

$$Probability\;(\mu\;-\;3\sigma\;\leqslant\;y\;\leqslant\mu\;+\;3\sigma)\;\geqslant\;(1-\frac{1}{K^2})\longrightarrow(1-\frac{1}{9})\longrightarrow(\frac{8}{9})\longrightarrow\;88.9\%$$


- so for 4th standard deviation i.e. k = 4 in Chebyshev equation: 

$$Probability\;(\mu\;-\;4\sigma\;\leqslant\;y\;\leqslant\mu\;+\;4\sigma)\;\geqslant\;(1-\frac{1}{K^2})\longrightarrow(1-\frac{1}{16})\longrightarrow(\frac{15}{16})\longrightarrow\;93.75\%$$

**NOTE: There is no 1st Standard Deviation for Y not belonging to gaussian distribution**

### Pareto Distrubtion

Pareto is transformed into Normal Distribution bcoz most algorithms assume data to be gaussian distributed and make good assumptions on it.

**Pareto is converted into Gaussian through following ways:**

Pareto Distribution $x\;\;=\;\;{x_1,\;x_2,\;x_3,\;x_4......x_n}$
                    
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\Downarrow$

Gaussian Distribution $y\;\;=\;\;{y_1,\;y_2,\;y_3,\;y_4......y_n}$

1. Apply a BoxCox function which will take all the x values. This will create a value $\alpha$ (Height of the Pareto graph)


2. Based on $\alpha\;value$, 'y' will be calculated

    - $y_i\;=\;\frac{(x_i)^\alpha\;-\;1}{\alpha}\;\;\;\;\; if\;\alpha\neq0\;\Longrightarrow\;BoxCox\;Transform$
    <br></br>
    - $y_i\;=\;log_e(x)\;\;\;\; if\;\alpha\;=\;0\;\Longrightarrow\;LOG\;Normal\;Distribution$

```python
# code to show BOX-COX Transformation of non-normal data

import numpy as np
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# generate non-normal data (exponential)
# exponential function is used to create Pareto distribution

original_data = np.random.exponential(size = 1000)

# transform training data and save lambda(alpha) value which will determines the height
# Passing the Pareto distribution to the boxcox transformation for the transformation

fitted_data, fitted_lambda = stats.boxcox(original_data)

# creating axes to draw plots
fig, ax = plt.subplots(1, 2)

# plotting the original data (non-normal)
sns.distplot(original_data, hist = False, kde = True,
            kde_kws = {'shade':True, 'linewidth':2},
            label = "Non-Normal", color = "green",ax = ax[0])

# plotting the fitted date (normal)
# fitted data is after performing boxcox transform over original exponential data
sns.distplot(fitted_data, hist = False, kde = True,
            kde_kws = {'shade':True, 'linewidth':2},
            label = "Normal", color = "green",ax = ax[1])

# adding legends to the subplots
plt.legend(loc = "upper right")

# rescaling the subplots
fig.set_figheight(5)
fig.set_figheight(10)

print(f"Lambda value used for Transformation: {fitted_lambda}")
```

```python
sm.qqplot(fitted_data, line = '45')
py.show()
```

It is not Gaussian Distributed

```python
# applyig log to the original data

sm.qqplot(np.log(original_data), line ='45')
py.show()
```

## Anova Test(F-Test)
The t-test works well when dealing with two groups, but sometimes we want to compare more than two groups at the same time.


For example, if we wanted to test whether petal_width age differs based on some categorical variable like species, we have to compare the means of each level or group the variable

### One Way F-test(Anova) :-
It tell whether two or more groups are similar or not based on their mean similarity and f-score.

Example : there are 3 different category of iris flowers and their petal width and need to check whether all 3 group are similar or not

**Step 4: Decision rule will be between df_between and df_within**

- df_between is numerator and df_within is denominator

- so from f-table, critical value = 3.5546 for df_between = 2 and df_within = 18 with $\alpha = 0.05$

- critical value =  3.55


**Step 5 : calculate F-test (https://www.youtube.com/watch?v=-yQb_ZJnFXw)**

- **Sum of Squares within groups ( SSW )** is the sum of variance across each seperate columns
    - calculate mean of each columns seperately
    	
                            0mg	50mg	100mg
        
        mean of each column	8.142857143	6.714285714	3

    
    - mean of each columns - each values
    - square of (mean of each columns - each values)
    - sum of these values
    - sum of square within group = **10.28571429**


- **Sum of Squares between groups** is the sum of variance beteen each groups
    - mean of each column - mean of total values from SST
    - (mean of each column - mean of total values from SST ) ^ 2
    - sum all values
    - sum of all values * (number of rows)
    - 14.0952381 * 7 = **98.66666667**


- **Total sum of Squares ( SST ) = Sum of squares between groups + Sum of Squares within groups**  
    - **Calculating Total Sum of Squares ( SST )**
        - put all values together and calculate the mean,  then individual value - mean
        - now, calculate (individual - mean)^2 and calculate total
        - This total is SST = 108.952381
        
$$Total\;sum\;of\;Squares\;( SST )\;\;=\;\;Sum\;of\;squares\;within\;groups\;\;+\;\;Sum\;of\;Squares\;between\;groups$$
        


- **Final calculation**
    
        - Sum of squares between groups
    
$$\frac{Sum\;of\;Squares\;between\;Groups}{degree\;of\;freedom}\Longrightarrow\;\frac{98.66}{2}\Longrightarrow\;49.33$$
                
                - where degree of freedom = number of columns - 1 = 3-1 = 2

        - Sum of squares between groups
    
$$\frac{Sum\;of\;Squares\;within\;Groups}{degree\;of\;freedom}\Longrightarrow\;\frac{10.28}{18}\Longrightarrow\;0.571$$
                
                - where degree of freedom = number of values in the table - number of columns = 21 - 3 = 18
                
$$F =  \frac{49.33}{0.571}\Longrightarrow 86.39$$

#### Step : 6
$$F(86.39)\;>\;critical\;value(3.55)$$
$$We\;reject\;null\;hypothesis$$
