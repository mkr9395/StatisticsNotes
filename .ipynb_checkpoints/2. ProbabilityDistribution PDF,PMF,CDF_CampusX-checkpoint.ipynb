## Probability

Probability is the chance of something happening, a more academic definition for this would be __the likelihood of an event occurring.__

An event is a specific outcome or a combination of several outcomes. These outcomes can be pretty much anything.

Having a probability of one expresses absolute certainty of the event occurring and a probability of zero expresses absolute certainty of the event not occurring. You probably figured this out, but higher probability values indicate a higher likelihood.

$ A = event $

$ P(A) = probability $
              
$$P(A) = \frac{Preferred\;(favourable)}{All\;(sample\;space)}$$


**The probability of two independent events occurring at the same time is equal to the product of all the probabilities of the individual events.**

$$ P(A\;and\;B) = P(A) . P(B)$$

eg: The likelihood of getting the ace of spades = (the probability of getting an ace) **times** (the probability of getting a spade.)

**Q. Why do we express probabilities numerically?**

Our goal is to be able to compare the probabilities of events and determine which is the relatively more likely outcome. 

**Experimental Probabilities :** The probabilities we get after conducting experiments.
whereas the ones we introduced earlier were theoretical or true probabilities.

The experimental probabilities we get are not always equal to the theoretical ones, but are a good approximation.
       
$$P(A) = \frac{Successful\;trials}{All\;trials}$$

**Q. Why do we use experimental probabilities?**

Because they are easy to compute and serve as good predictors for theoretical ones

**Expected value E(A) :** It is used to predict future events. It is The average outcome we expect if we run an experiment many times.

The expected value of an event A denoted **E(A)** is the outcome we expect to occur when we run an experiment.

We can use expected values to make predictions about the future based on past data. 

eg: *We frequently make predictions using intervals instead of specific values due to the uncertainty the future brings.*
Meteorologists often use these when forecasting the weather, they do not know exactly how much snow,rain or wind there's going to be, so they provide us with likely intervals That is why we often hear statements like expect between three to five feet of snow tomorrow morning or temperatures rising up to 90 degrees on Wednesday.

**For Categorical Outcome :**

For an event with categorical outcomes like suits, we calculate the expected value by multiplying the theoretical probability of the event P(A)  by the number of trials we carried out in (n).

**E(A) = P(A) * n**

**Numerical outcome :**

**A + B + C =  P(A) . A +  P(B) . B +  P(C) . C**

We take the value for every element in the sample space and multiply it by its probability, then we add all of those up to get the expected value.

**Q. Why do we use intervals when forecasting future events?**

1. because the expected value might have a low probability of occuring.
2. because we want to increase the likelihood of our predictions being accurate.
3. because the expected value could not be attainable.

### 1. __Probability Mass Function (PMF) :__


- It is a mathematical function that describes the probability distribution of a ___discrete random variable.___


- The PMF of a discrete random variable assigns a probability to each possible value
of the random variable. 


___The probabilities assigned by the PMF must satisfy two
conditions:___

- a. The probability assigned to each value must be non-negative (i.e., greater
than or equal to zero).


- b. The sum of the probabilities assigned to all possible values must equal 1

CDF describes the probability that a random variable $'X'$ with a given probability distribution will be found at a __value less than or equal to $'x'$__

$$f(x)\;=\;P\;(X<=x)$$

### Difference between Probability and Probability Density:

The main difference between probability and probability density is:

- **Probability**: It measures the likelihood of a specific outcome or event occurring in a discrete sample space. It gives a probability value for each specific outcome and is used for discrete random variables.



- **Probability Density**: It measures the likelihood of a continuous random variable taking on a particular value or falling within a specific range. It gives a probability density value for each point on the continuous spectrum and is used for continuous random variables.

In summary, probability deals with discrete events and assigns probabilities to specific outcomes, while probability density deals with continuous variables and assigns probabilities to intervals or ranges of values.




**Example - Rolling a Die**:

**Probability**:
- Suppose you have a standard six-sided die, and you want to find the probability of rolling a 3. In this case, you have a discrete set of possible outcomes: {1, 2, 3, 4, 5, 6}.
- The probability of rolling a 3 is 1/6, because there is 1 favorable outcome (rolling a 3) out of a total of 6 possible outcomes.
- Probability deals with specific events or outcomes, and it assigns probabilities to each outcome individually.

**Probability Density**:
- Now, consider a different scenario where you want to find the probability density of rolling a die and obtaining a result between 3 and 4 (inclusive) on a continuous scale.
- In this case, you are working with continuous values within a range: [3, 4].
- The probability density function (PDF) for this continuous interval is constant because each value within the interval has an equal likelihood of occurring.
- The probability density within this range would be 1/(4 - 3) = 1, indicating a constant density for this interval.
- Probability density deals with continuous variables and assigns probabilities to intervals or ranges of values, rather than individual specific outcomes.

In summary:

- **Probability** deals with discrete events and assigns probabilities to specific outcomes (e.g., the probability of rolling a 3 on a die).
- **Probability Density** deals with continuous variables and assigns probabilities to intervals or ranges of values (e.g., the probability density of rolling a die and getting a result between 3 and 4 on a continuous scale).

### Density Estimation - use to calculate PDFs

Density estimation is a statistical technique used to estimate the probability density function (PDF) of a random variable based on a set of observations or data. In simpler terms, it involves estimating the underlying distribution of a set of data points.


Density estimation  is often used to estimate the probability distribution of input data or to model the likelihood of certain events or outcomes.


There are various methods for density estimation, __including parametric and non-parametric approaches.__ 

- __Parametric methods__ assume that the data follows a specific probability distribution (such as a normal distribution), while 


- __Non-parametric methods__ do not make any assumptions about the distribution and instead estimate it directly from the data.

<br></br>

Commonly used techniques for density estimation include 
- __Kernel Density Estimation (KDE)__ 
- __Histogram estimation__
- __Gaussian mixture models (GMMs)__

### 1. Parametric Density Estimation
Parametric density estimation is a method of estimating the probability density
function (PDF) of a random variable by assuming that the underlying distribution
belongs to a specific parametric family of probability distributions, such as the
normal, exponential, or Poisson distributions.

___It is called Parameteric as we plot distribution using parameters such as mean and std___

### Non-Parametric Density Estimation (KDE)

But sometimes the distribution is not clear or it's not one of the famous distributions.


- Non-parametric density estimation is a statistical technique used to estimate the probability
density function of a random variable without making any assumptions about the underlying
distribution. 


- It is also referred to as non-parametric density estimation because __it does not
require the use of a predefined probability distribution function, as opposed to parametric
methods such as the Gaussian distribution.__


- The non-parametric density estimation technique involves constructing an estimate of the
probability density function using the available data. This is typically done by creating a __kernel
density estimate.__

##### Advantage of Non-parametric density estimation over parametric density estimation:

- One of the main advantages is that it does not require the assumption of a specific distribution, which allows for more flexible and accurate estimation in situations where the underlying distribution is unknown or complex. 

##### Disadvantage of Non-parametric density estimation over parametric density estimation:
- Non-parametric density
estimation can be computationally intensive and may require more data to achieve accurate
estimates compared to parametric methods.

### Kernel Density Estimate(KDE)

The KDE technique involves using a __kernel function__ to smooth out the data and create a
continuous estimate of the underlying density function.

here we are using gaussian kernel: 

- ___considering each data point as centre(or mean) and create a Gaussian kernal around it.___


- ___moving on y-axis and summing proability denisities of each peak of Gaussian kernal which has same x-points.___

### steps to calculate KDE:

 Suppose we have the following dataset of ages: [25, 27, 30, 33, 35, 36].

Step 1: Data Preparation
- Start with the dataset: Ages = [25, 27, 30, 33, 35, 36].

Step 2: Choose a Kernel Function
- Select a kernel function. For this example, we'll use a Gaussian (normal) kernel, which is commonly used in KDE. The Gaussian kernel is defined as: K(u) = (1 / √(2π)) * e^(-u^2 / 2), where u is a variable.

Step 3: Choose a Bandwidth (Smoothing Parameter, like $\sigma$ in case of normal distribution) 
- Select a bandwidth parameter (h). Let's choose h = 2 for this example. This bandwidth determines the width of the kernel and influences the smoothness of the estimate.

Step 4: Calculate Kernels for Each Data Point
- For each data point in the dataset, we apply the Gaussian kernel centered at that point with the chosen bandwidth. Let's calculate the kernels for each data point:

   - For the data point 25:
     K(x) = (1 / √(2π)) * e^(-(x - 25)^2 / (2 * 2^2)) = 0.054

   - For the data point 27:
     K(x) = (1 / √(2π)) * e^(-(x - 27)^2 / (2 * 2^2)) = 0.119

   - Continue this calculation for each data point.

Step 5: Sum the Kernels
- Sum all the individual kernels calculated in Step 4 for each data point. This results in a continuous curve representing the estimated probability density.

   - At age 25, the density is 0.054.
   - At age 27, the density is 0.119.
   - Continue this process for all data points.

Step 6: Normalize the Result
- To ensure that the estimated density integrates to 1, divide the entire curve by the total number of data points (6) and the bandwidth (2).

Step 7: Plot the KDE Curve
- Finally, plot the resulting KDE curve on a graph to visualize the estimated probability density of the ages.

The plotted KDE curve will provide a smooth representation of how ages are likely to be distributed across the age range, with higher densities indicating where data points are more concentrated.

In this example, we used a Gaussian kernel with a bandwidth of 2 to estimate the probability density of ages based on the given dataset. The KDE curve will show the shape and smoothness of the estimated age distribution.

```python
# generate a sample
sample1 = normal(loc=20, scale=5, size=300)
sample2 = normal(loc=40, scale=5, size=700)
sample = np.hstack((sample1, sample2))
```

#### graph not following any particular distribution so we will use KDE (non-parametric)

#### Applying KDE and kernal as Gaussian :


```python
from sklearn.neighbors import KernelDensity

# bandwidth gives spread (std dev) of the graph. 

model = KernelDensity(bandwidth=3, kernel='gaussian') # by deafult it is gaussian
```

### NOTE : higher bandwidth gives flatter graphs and lower bandwidth give peaked graphs

```python
# convert data to a 2D array for KDE to work
sample = sample.reshape((len(sample), 1))

model.fit(sample)
```

##### genearting 100 points between min and max

```python
values = np.linspace(sample.min(),sample.max(),100)
values = values.reshape((len(values), 1)) # converting those 100 points to 2-D
```

```python
probabilities = model.score_samples(values) # will give log values
```

```python
# calculating exponential value of each probability as it converts log values to normal

probabilities = np.exp(probabilities)
```

`score_samples(values)` returns the log-density estimate of the input samples values. 

This is because the `score_samples()` method of the KernelDensity class returns the logarithm of the probability density estimate rather than the actual probability density estimate.

```python
plt.hist(sample, bins=50, density=True)
plt.plot(values[:], probabilities)
plt.show()
```

### NOTE

- __Area of PDF gives CDF (Integration on PDF gives CDF)__


- __slope of CDF for each point makes the PDF (Differentiation on CDF gives PDF)__

CDF is especially useful when we want to estimate the probability of some interval.Graphically, the area under the density curve would represent the chance of getting a value within that interval. We find this area by computing the integral of the density curve over the interval from A to B.
$$\int_a^b p(x)dx$$




$$PDF \longrightarrow CDF \int_ {-\infty}^y p(y)dy = F(y) \Longrightarrow Integration\;\text{(Area under curve of PDF)}$$ 
<br></br>
$$CDF \longrightarrow PDF\;\;p(y)=F(y)\frac{d}{dy}  \leftarrow Derivative\;(Slope\;of\;CDF\;using\;Differentitation)$$
