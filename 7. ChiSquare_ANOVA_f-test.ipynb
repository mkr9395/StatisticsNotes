## Chi-square distribution $(\chi ^2)$

- The Chi-Square distribution, is a __continuous
probability distribution__ that is widely used in statistical hypothesis testing,
particularly in the context of<br></br>
     - __goodness-of-fit tests__ and <br></br>
     - __tests for independence__ in contingency tables. 


- The Chi-Square distribution has a single parameter, the degrees of freedom (df),
which influences the shape and spread of the distribution. 


$$Denoted\;by\;\; \chi^2*(k)$$
$$Y \sim\;\chi^2*(k)$$

   we denote chi square distribution as the capital Greek letter $\chi$ squared, followed by a parameter k depicting the degrees of freedom.



- __It is taking sample from a normal distribution and squaring it. The KDE of it is chi-sqaure$(\chi ^2 = N^2)$__<br></br>


- __The degree of
freedom are typically associated with the number of independent variables or
constraints in a statistical problem.__ i.e how many normal distributions we are squaring and adding together.<br></br> 
   - df = 1  sum of sqaure of single Normal distribution<br></br>
   
   - df = 2 when we perform sum of sqaure of two Normal distrutions<br></br>
   
   - df = 3 when we perform  sum of sqaure of three Normal distrutions


__for very high degree of freedom, chi-square distribution becomes similar to NORMAL DISTRIBUTION__

### NOTE : As degree of freedom increases, the graph becomes more right skewed

### __Some key properties of the Chi-Square distribution are:__
- It is a continuous distribution, defined for non-negative values.


- It is positively skewed, with the degree of skewness decreasing as the
degrees of freedom increase.


- __The mean of the Chi-Square distribution is equal to its degrees of
freedom, and its variance is equal to twice the degrees of freedom.__
    - $\mu = k$
    - $\sigma = \sqrt{2k}$ where k is degree of freedom


- As the degrees of freedom increase, the Chi-Square distribution
approaches the normal distribution in shape.


The Chi-Square distribution is used in various statistical tests, such as 

- __the Chi- Square goodness-of-fit test__, which evaluates whether an observed frequency 
distribution fits an expected theoretical distribution, and 


- the __Chi-Square test for independence__, which checks the association between categorical variables in a
contingency table.

### 1. Goodness of Fit Test

The Chi-Square Goodness-of-Fit test is a statistical hypothesis test used to
determine if the observed distribution of a single categorical variable
matches an expected theoretical distribution. 


It helps to evaluate
whether the data follows a specific probability distribution, such as
uniform, binomial, or Poisson distribution, among others. This test is
particularly useful when you want to assess if the sample data is
consistent with an assumed distribution or if there are significant
deviations from the expected pattern.

### Steps: 


The Chi-Square Goodness-of-Fit test involves the following steps:

1. ___Decide which distribution the theoretical values will follow and calculate the theoretical values accordingly___


2. __Define the null hypothesis (H0) and the alternative hypothesis (H1):__
    - H0: The observed data follows the expected theoretical distribution.<br></br>
    
    - H1: The observed data does not follow the expected theoretical distribution.
<br></br>

3. __Calculate the theoretical frequencies (ideally how much should have come) for each category based on the
theoretical distribution and the sample size.__


4. __Compute the Chi-Square test statistic (χ2) by comparing the theoretical
and observed frequencies.__ 
 

   
### The test statistic is calculated as:


$$\text{Formula for Chi-Square test Statistics :}$$

$$\chi^2 = \sum{\frac{(f_o\;-\;f_e)^2}{f_e}}$$


$where\;f_o\longrightarrow\;\text{frequency of observed}$

$where\;f_e\longrightarrow\;\text{frequency of theoretical}$

5. __where $f_o$ is the observed frequency in category o, $f_e$ is the theoretical
frequency in category e, and the summation is taken over all
categories.__


6. __Determine the degrees of freedom (df), which is typically the
number of categories minus one (df = k - 1), where k is the number
of categories.__


7. __Calculate the p-value for the test statistic using the Chi-Square
distribution with the calculated degrees of freedom.__


8. __Compare the test statistic to the critical value or the p-value__

### Assumptions

1. __Independence:__ The observations in the sample must be independent
of each other. This means that the outcome of one observation
should not influence the outcome of another observation.


2. __Categorical data:__ The variable being analysed must be categorical,
not continuous or ordinal. The data should be divided into mutually
exclusive and exhaustive categories.


3. __Expected frequency:__ Each category should have an expected frequency of at least 5. This guideline helps ensure that the Chi-Square distribution is a reasonable approximation for the distribution of the test statistic. Having small expected frequencies can lead to an inaccurate estimation of the Chi-Square distribution,
    potentially increasing the likelihood of a Type I error (incorrectly
    rejecting the null hypothesis) or a Type II error (incorrectly failing to
    reject the null hypothesis).
    


4. __Fixed distribution:__ ___The theoretical distribution being compared to
the observed data should be specified before the test is conducted.___
It is essential to avoid choosing a distribution based on the observed
data, as doing so can lead to biased results.

### __NOTE :  The Chi-Square Goodness-of-Fit test is a  non-parametric test.__ 

Non-
parametric tests do not assume that the data comes from a specific probability distribution or make any assumptions about population
parameters like the mean or standard deviation.


In the Chi-Square Goodness-of-Fit test, we compare the observed
frequencies of the categorical data to the expected frequencies based on
a hypothesized distribution. The test doesn't rely on any assumptions
about the underlying distribution's parameters. Instead, it focuses on comparing observed counts to expected counts, making it a non-
parametric test.

### Example 1 (godness of fit): 

Suppose we have a six-sided fair die, and we want to test if the die is indeed fair. We roll the
die 60 times and record the number of times each side comes up. We'll use the Chi-Square
Goodness-of-Fit test to determine if the observed frequencies are consistent with a fair die
(i.e., a uniform distribution of the sides).


Observed frequencies for all values:
- ○ Side 1: 12 times
- ○ Side 2: 8 times
- ○ Side 3: 11 times
- ○ Side 4: 9 times
- ○ Side 5: 10 times
- ○ Side 6: 10 times

- __we will assume that the fair dice follows a normal distribution__


- ${H_o}$ : die is fair



- $H_1$: die is not fair


- $\alpha=0.05$



- __Observed values $f_0$__: sum of all outcomes = 12 + 8 + 11 + 9 + 10 + 10 = 60


- __Expected/theoretical values $f_e$__ : as it follows normal distribution so probability of each outcome will be $60 * \frac{1}{6}=10\;times$

<br></br>
$$\chi^2=\frac{(12-10)^2+(8-10)^2+(11-10)^2+(9-10)^2+(10-10)^2+(10-10)^2}{10}=\frac{10}{10}=1$$

$$(test-statistics)\longrightarrow \chi^2\;=1$$

<br></br>
number of categories = 6 

- __so degree of freedom =__ $n-1=6-1=5$

since $\alpha=0.05$ and p-value = 0.9625. 

__so we cannot reject the null hypothesis as p-value > 0.05. So it is a fair die__

#### using python : 

```python
import numpy as np
import scipy.stats as stats
```

```python
# Observed frequencies
observed_frequencies = np.array([12, 8, 11, 9, 10, 10])
observed_frequencies
```

```python
# Expected frequencies for a fair die (uniform distribution)
expected_frequencies = np.array([10, 10, 10, 10, 10, 10])
expected_frequencies
```

```python
# Calculate the chi-square statistic
chi_square_statistic = np.sum((observed_frequencies - expected_frequencies) ** 2 / expected_frequencies)
chi_square_statistic
```

```python
# Degrees of freedom (number of categories - 1)
degrees_of_freedom = len(observed_frequencies) - 1
degrees_of_freedom
```

```python
# Set the significance level (alpha)
alpha = 0.05
```

```python
# Calculate the critical value from the chi-square distribution
critical_value = stats.chi2.ppf(1 - alpha, degrees_of_freedom)
critical_value
```

####  NOTE : `1 - stats.chi2.cdf(chi_square_statistic, degrees_of_freedom)` 

#### and `stats.chi2.sf(chi_square_statistic, degrees_of_freedom)` are same.  i.e `chi2.sf = 1-chi2.cdf`

```python
# Calculate the p-value
p_value = 1 - stats.chi2.cdf(chi_square_statistic, degrees_of_freedom)
p_value
```

```python
# same p-value
stats.chi2.sf(chi_square_statistic, degrees_of_freedom)
```

```python
# Print the results
print(f"Chi-Square Statistic: {chi_square_statistic}")
print(f"Critical Value: {critical_value}")
print(f"P-value: {p_value}")
```

```python
# Check if the null hypothesis is rejected
if chi_square_statistic > critical_value:
    print("Reject the null hypothesis. The die is not fair.")
else:
    print("Fail to reject the null hypothesis. The die appears to be fair.")
```

## 2. Test for Independence : 2 columns are independent

The Chi-Square test for independence, also known as the Chi-Square test for association, is a
statistical test used to determine whether there is a significant association between two
categorical variables in a sample. It helps to identify if the occurrence of one variable is
dependent on the occurrence of the other variable, or if they are independent of each other.


The test is based on comparing the observed frequencies in a contingency table (a table that
displays the frequency distribution of the variables) with the frequencies that would be
expected under the assumption of independence between the two variables.

### NOTE : Here degree of freedom: df = (number of rows - 1) * (number of columns - 1)

### Steps : 

1. State the null hypothesis (H0) and alternative hypothesis (H1):
    - H0: There is no association between the two categorical variables (they are
independent).

    - H1: There is an association between the two categorical variables (they are
dependent).


2. Create a contingency table with the observed frequencies for each combination of the
categories of the two variables.



3. Calculate the expected frequencies for each cell in the contingency table assuming that
the null hypothesis is true (i.e., the variables are independent).


4. __Compute the Chi-Square test statistic:__
$$\chi^2 = Σ \frac {(f_0 - f_e)^2}{f_e}$$

- $f_o$__: is the observed frequency in each cell__ 


-  $f_e$__: is the expected frequency.__


5. __Determine the degrees of freedom: df = (number of rows - 1) * (number of columns - 1)__



6. Obtain the critical value or p-value using the Chi-Square distribution table or a statistical
software/calculator with the given degrees of freedom and significance level (commonly
α = 0.05).


7. Compare the test statistic to the critical value or the p-value to the significance level to
decide whether to reject or fail to reject the null hypothesis. If the test statistic is greater
than the critical value, or if the p-value is less than the significance level, we reject the null
hypothesis and conclude that there is a significant association between the two variables.

### Assumptions
1. __Independence of observations:__ The observations in the sample should be independent of
each other. This means that the occurrence of one observation should not affect the
occurrence of another observation. In practice, this usually implies that the data should
be collected using a simple random sampling method.


2. __Categorical variables:__ Both variables being tested must be categorical, either ordinal or
nominal. The Chi-Square test for independence is not appropriate for continuous
variables.


3. __Adequate sample size:__ The sample size should be large enough to ensure that the
expected frequency for each cell in the contingency table is sufficient. A common rule of
thumb is that the expected frequency for each cell should be at least 5. If some cells have
expected frequencies less than 5, the test may not be valid, and other methods like
Fisher's exact test may be more appropriate.


4. __Fixed marginal totals:__ The marginal totals (the row and column sums of the contingency
table) should be fixed before the data is collected. This is because the Chi-Square test for
independence assesses the association between the two variables under the assumption
that the marginal totals are fixed and not influenced by the relationship between the
variables.

### Example 1

A researcher wants to investigate if there is an association between the level of education
(categorical variable) and the preference for a particular type of exercise (categorical variable)
among a group of 150 individuals. The researcher collects data and creates the following
contingency table




- ${H_o}$ : education and exercise are independent



- $H_1$: education and exercise are not independent



- $\alpha=0.05$



- __Observed values $f_0$__: 




### NOTE : for expected value table we will multiply (probabilities of both the respective totals of column and row happening/ Grand total) * Grand total

eg: for going to high school and doing yoga we will do high school total = 45 and yoga total = 40. The grand total = 150

$$\frac{\text{going to high school}}{\text{grand total}} * \frac{\text{doing yoga}}{\text{grand total}}* \text{Grand total}$$

$$\frac{45}{150}*\frac{40}{150}*150=12$$

$$OR$$


$$\frac{\text{going to high school * doing yoga}}{\text{grand total}} \Longrightarrow \frac{45 * 40}{150}=12$$

- __Expected values $f_e$__ : 




$$\chi^2=\frac{(15-12)^2}{12}+\frac{(20-19)^2}{19}.............=9.95(assumed)$$


<br><br>
- number of rows = 3

- number of columns = 3



- __degree of freedom__

    - $(3-1)*(3-1) = 4$

```python
import scipy.stats as stats

test_statistic = 9.95  # Chi-Square test statistic from the previous example
degrees_of_freedom = 4  # Degrees of freedom from the previous example

# Calculate the p-value using the chi2 survival function (sf)
p_value = stats.chi2.sf(test_statistic, degrees_of_freedom)

print("P-value:", p_value)
```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nWe reject the null hypothesis. Hence, education and exercise are not independent. there is some association between them")
else:
    print("\nWe fail to reject the null hypothesis. Hence, education and exercise are independent")
```

#### use of np.outer : 

`np.outer` is a function in the NumPy library that computes the outer product of two arrays. The outer product is a mathematical operation that takes two vectors (or arrays) and returns a matrix as the result. The resulting matrix has dimensions determined by the lengths of the input vectors.

- `a` and `b` are input arrays or vectors.


- The function returns a matrix that represents the outer product of `a` and `b`.

The outer product is calculated as follows:

1. For each element `a[i]` in the first input array `a` and each element `b[j]` in the second input array `b`, the function computes the product `a[i] * b[j]`.

2. These products are then arranged in a matrix, where the element at position `(i, j)` in the resulting matrix corresponds to the product of `a[i] * b[j]`.

The resulting matrix is often used in various mathematical and computational operations, including linear algebra, statistics, and machine learning.

```python
import numpy as np

a = np.array([1, 2, 3])
print(a)
b = np.array([4, 5, 6])
print()
print(b)
print()

outer_product = np.outer(a, b)

print(outer_product)
```

Here's why we need ANOVA even when t-tests are available:

1. **Comparison of Multiple Groups:**<br></br>

   - T-tests, on the other hand, are designed for comparing the __means of two groups.__ If you have more than two groups, conducting multiple t-tests for pairwise comparisons can lead to an increased risk of Type I errors (false positives).



2. **Type I Error Control:**

### One way ANOVA test

- One-way ANOVA (Analysis of Variance) is a statistical method used to __compare the means of
three or more independent groups__ to determine if there are any significant differences
between them. It is an extension of the t-test, which is used for comparing the means of two
independent groups.


- The term "one-way" refers to the fact that there is __only one independent
variable (factor) with multiple levels (groups)__ in this analysis. eg : employee column with  analyst, senior analyst and manager category



- The primary purpose of one-way ANOVA is to test the ___null hypothesis that all the group means
are equal.___ The __alternative hypothesis__ is that at least one group mean is significantly different
from the others.

## Steps : 

### 1. Define the null and alternative hypotheses.

- $H_0 :$ mean of all the groups are equal


- $H_1 :$ mean of atleast 1 group is significantly different

### 2. Calculate the overall mean (grand mean) of all the groups combined and mean of all the groups individually.

$$\text{grand mean = mean of all values combined}\;(\bar{\bar x})\;=\;6$$

### 3. Calculate the "sum of square total",  "between-group" and "within-group" sum of squares (SS).

#### a. calculating SST (Sum of squares total) : 

> __This gives overall variance of the data. distance between each point with the Grand Mean.__

$$\sum (Grand\;total\;mean\; - \;each\;element)^2$$

- also called as variance of entire dataset



##### degree of freedom for SST = n -1 = 9 - 1 = 8 (not required to be calculate, not used anywhere)

#### b. calculating SSW (Sum of squares within samples or categories) : 


$$\sum (mean\;of\;each\;group - elements\;of\;that\;group)^2$$

> __Internal variance of each group__

##### degree of freedom of each group = ( n - k )

total degree of freedom = n - k = 9 - 3 = 6

#### c. calculating SSB (Sum of squared between samples or categories) and degree of freedom : 


$$\sum (\text{number of sample in that group}*\;(Grand\;mean\;-\;mean\;of\;each\; group)^2)$$

> __variance between grand mean and each group mean__

##### - degree of freedom = (k - 1) = 3 - 1 = 2

#### finally :




## Rule : SST = SSW + SSB (for both values and df)

### 4. F-ratio :

#### Calculate the "between-group" and "within-group" mean squares (MS) by dividing their respective sum of squares by their degrees of freedom.

$$a.\;\frac{SS_{between}}{df_{SSB}} = \frac{24}{2}=12$$



$$b.\;\frac{SS_{within}}{df_{SSW}} = \frac{52}{6}$$

### 5. Calculate the F-statistic by dividing the "between-group" mean square by the "within- group" mean square.

$$\text{f-statistics}\Longrightarrow\frac{SS_{between}}{df_{SSB}} * \frac{df_{SSW}}{SS_{within}}$$
<br></br>


$$F=\frac{24}{2} * \frac{6}{52} = \frac{18}{13} \approx 1.4 $$
<br></br>
$$\text{So, f-statistics} = 1.4$$

### 6. Calculate the p-value associated with the calculated F-statistic using the F-distribution and the appropriate degrees of freedom. 


- The p-value represents the probability of obtaining
an F-statistic as extreme or more extreme than the calculated value, assuming the null
hypothesis is true.


- Choose a significance level (alpha), typically 0.05.


- Compare the calculated p-value with the chosen significance level (alpha).
    
    - If the p-value is less than or equal to alpha, reject the null hypothesis in favour of the alternative
hypothesis, concluding that there is a significant difference between at least one pair of group
means.
    - If the p-value is greater than alpha, fail to reject the null hypothesis, concluding that there is not
enough evidence to suggest a significant difference between the group means.

#### calculating p-value for given f-statistic:

```python
import scipy.stats as stats

f_statistic = 1.4  # The F-statistic value you've calculated
df1 = 2            # Degrees of freedom for the numerator (between groups)
df2 = 6           # Degrees of freedom for the denominator (within groups)

p_value = stats.f.sf(f_statistic, df1, df2)
print("P-value:", p_value)
```

### Conclusion : since p-value 0.31 > significance level ($\alpha$) 0.05 . So NULL Hypothesis can't be rejected 


$$Hence\; \mu_{A}=\mu_{B}=\mu_C$$

- It's important to note that one-way ANOVA only determines if there is a significant difference
between the group means


- it does not identify which specific groups have significant differences. 


- To determine which pairs of groups are significantly different, post-hoc tests, such
as Tukey's HSD or Bonferroni, are conducted after a significant ANOVA result.

### Points to remember : 


- __F-statistics calculated will follow f-distribution as SSB and SSW will follow chi-square distribution.__



- __F-statisctics will always between 0 to +ve, it can never be -ve.__


- __F$\uparrow$ when SSB $\uparrow$ SSW $\downarrow$ ie. sample mean (SSB) will be very far from parent population mean. Mean of all 3 will never be same. Hence p-value will be small.__


- __F$\downarrow$ when SSB $\downarrow$ SSW $\uparrow$ i.e graph will be flatter. All child population will be very close to each other and to the parent population. Mean will be same to each other.__ 



-  __we are not proving that  $\mu_A=\mu_B=\mu_C$ which is the null hypothesis, insted we are proving that $\mu_A, \mu_B, \mu_C$ are not different population but are part  of the same larger population. Hence population mean will be same for all.__



-  __$\mu_A, \mu_B, \mu_C$ are mean of 3 distributions from the same population. The population not necesaarily will be normal distribution__




#### Intuition of SSB $\downarrow$

- All child population will be very close to each other and to the parent population. Mean will be same to each other.



### Assumptions
1. __Independence :__ The observations within and between groups should be independent of
each other. This means that the outcome of one observation should not influence the
outcome of another. Independence is typically achieved through random sampling or
random assignment of subjects to groups.


2. __Normality :__ The data within each group should be approximately normally distributed.
While one-way ANOVA is considered to be robust to moderate violations of normality,
severe deviations may affect the accuracy of the test results. If normality is in doubt, non-
parametric alternatives like the Shapiro-wilk test can be considered.



3. __Homogeneity of variances :__ The variances of the populations from which the samples are
drawn should be equal, or at least approximately so. This assumption is known as
homoscedasticity. If the variances are substantially different, the accuracy of the test
results may be compromised. Levene's test or Bartlett's test can be used to assess the
homogeneity of variances. If this assumption is violated, alternative tests such as Welch's
ANOVA can be used.

### Difference between one way anova and two way anova:

Two-way ANOVA (Analysis of Variance) and one-way ANOVA are both statistical techniques used to analyze and compare groups or datasets. However, they differ in terms of the number of factors or independent variables they consider and the types of questions they address:

1. **One-Way ANOVA:**
   - One-way ANOVA is used when you have one categorical independent variable (factor) with more than two levels or groups, and you want to compare the means of these groups to determine if there are statistically significant differences among them.
   - One-way ANOVA answers questions such as whether there are significant differences in test scores among students taught by different teachers or whether there are differences in the performance of three different drug treatments on a medical condition.
   - The factor in one-way ANOVA represents the categories or groups you are comparing, and you test whether their means are significantly different.



2. **Two-Way ANOVA:**
   - Two-way ANOVA, on the other hand, is used when you have two independent categorical variables (factors) and want to examine how they interact with each other and how they independently influence a dependent variable.
   - Two-way ANOVA is appropriate for situations where you want to answer questions like whether there are significant differences in test scores due to the teacher's teaching method (factor 1), the students' background (factor 2), and whether there is an interaction effect between the two factors.
   - In two-way ANOVA, you can assess the main effects of each factor (independently) and the interaction effect between the two factors.

Key differences between one-way ANOVA and two-way ANOVA:

   - One-way ANOVA involves one categorical factor.
   - Two-way ANOVA involves two categorical factors.

- **Main Effects and Interaction:**
   - One-way ANOVA tests for differences among the groups defined by a single factor.
   - Two-way ANOVA tests for differences due to two factors and assesses both main effects (independent influences of each factor) and interaction effects (combined influence of the factors).

- **Research Questions:**
   - One-way ANOVA answers questions about differences among groups defined by one factor.
   - Two-way ANOVA answers questions about how two factors independently and together influence a dependent variable.

In summary, the key distinction between one-way and two-way ANOVA is the number of independent categorical factors being considered and the types of research questions they address. One-way ANOVA is used for comparing groups defined by a single factor, while two-way ANOVA is used for examining the effects of two factors and their interactions on a dependent variable.

## Two-way ANOVA:


https://www.youtube.com/watch?v=cNIIn9bConY&ab_channel=statisticsfun

- Two-way ANOVA (Analysis of Variance) is a statistical technique used to analyze the influence of __two categorical independent variables (factors) on a continuous dependent variable.__


- It extends the concept of one-way ANOVA, which assesses the impact of a single factor. Two-way ANOVA is more complex but provides a more comprehensive analysis by considering interactions between the two factors. 



**Formula of Two-Way ANOVA:**

The general formula for a two-way ANOVA model with two factors (A and B) is as follows:

$$Y_{ij} = \mu + A_i + B_j + (A \cdot B)_{ij} + \varepsilon_{ij}$$

- $Y_{ij}$ is the observed value for the dependent variable.
- $\mu$ is the overall mean.
- $A_i$ represents the effect of factor A for level $i$.
- $B_j$ represents the effect of factor B for level $j$.
- $(A \cdot B)_{ij}$ represents the interaction effect between factor A at level $i$ and factor B at level $j$.
- $\varepsilon_{ij}$ is the random error term.

**Steps for Two-Way ANOVA:**

1. **Data Preparation:**
   - Collect and organize your data into a table or dataframe with columns for the dependent variable and two categorical independent variables (factors A and B).

2. **Hypotheses:**
   - Formulate null and alternative hypotheses:
     - Null Hypothesis (H0): There are no significant main effects of factors A and B, and there is no significant interaction effect between them.
     - Alternative Hypothesis (Ha): At least one main effect or interaction effect is significant.

3. **Calculate Group Means:**
   - Calculate the means of the dependent variable for each combination of factor levels (A and B).

4. **Calculate Sum of Squares (SS):**
   - Calculate the sum of squares for each factor (SSA for factor A, SSB for factor B) and the interaction effect (SSAB).
   - SS = $\sum (Y_{ij} - \mu)^2$
   - SSA, SSB, and SSAB represent the variability explained by each factor.

5. **Degrees of Freedom (df):**
   - Calculate the degrees of freedom for each source of variation (dfA, dfB, dfAB) based on the number of levels in each factor.

6. **Mean Squares (MS):**
   - Calculate mean squares for each source of variation by dividing SS by its respective degrees of freedom.

7. **F-Statistic:**
   - Calculate the F-statistic for each factor and interaction using the mean squares:
     - $F_A = \frac{MSA}{MSE}$
     - $F_B = \frac{MSB}{MSE}$
     - $F_{AB} = \frac{MSAB}{MSE}$
   - Where MSE is the mean square of error (residuals).

8. **P-Value:**
   - Calculate the p-values associated with the F-statistics using the F-distribution.

9. **Hypothesis Testing:**
   - Conduct hypothesis tests for each factor and interaction:
     - For factor A: Compare $F_A$ to the critical value or calculate the p-value.
     - For factor B: Compare $F_B$ to the critical value or calculate the p-value.
     - For interaction: Compare $F_{AB}$ to the critical value or calculate the p-value.

10. **Interpretation:**
    - If any of the F-tests for factors A, B, or the interaction are statistically significant (p < alpha), you reject the null hypothesis.
    - Determine the significance of the main effects and interaction effects.

**Differences from One-Way ANOVA:**

1. **Number of Factors:**
   - One-way ANOVA has one categorical factor, while two-way ANOVA has two factors.

2. **Interaction Effect:**
   - One-way ANOVA assesses only the main effect of the single factor.
   - Two-way ANOVA assesses the main effects of two factors and the interaction effect between them.

**Example:**

Consider an example in which you are analyzing the impact of two factors, "Teaching Method" (A) and "Student Background" (B), on students' test scores. In one-way ANOVA, you'd only analyze the effect of one of these factors, such as "Teaching Method." In two-way ANOVA, you simultaneously assess how both factors, "Teaching Method" and "Student Background," influence test scores and whether there's an interaction between them. This provides a more comprehensive understanding of the factors influencing test scores.

- In this example, we'll analyze the impact of two factors, "Teaching Method" and "Student Background," on students' test scores. We'll use a hypothetical dataset and perform a two-way ANOVA to understand how these factors affect test scores.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Load the Titanic dataset (assuming you have it in a CSV file)
# Replace 'titanic.csv' with the actual path to your dataset.
titanic_df = pd.read_csv(r"D:\Statistics\titanic\train.csv")
titanic_df.head()
```

```python
# Perform a two-way ANOVA
model = ols('Fare ~ C(Pclass) * C(Sex)', data=titanic_df).fit()
```

```python
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
# Print the ANOVA table
print(anova_table)
```

The ANOVA (Analysis of Variance) table provides statistical results for the analysis of variance, specifically for a two-way ANOVA in this case. The table includes several statistics and p-values that help you assess the significance of different effects in your model. Let's break down the components of the ANOVA table and how to interpret them:

1. **Source of Variation**: The table typically starts with a breakdown of the sources of variation. In your two-way ANOVA, there are three main sources:

   - `C(Pclass)`: This represents the variation due to the "Pclass" variable.
   - `C(Sex)`: This represents the variation due to the "Sex" variable.
   - `C(Pclass):C(Sex)`: This represents the interaction effect between "Pclass" and "Sex." It captures whether the combination of "Pclass" and "Sex" has an additional effect on the dependent variable ("Fare") beyond what you would expect from considering each factor individually.

2. **Sum of Squares (SS)**: The sum of squares represents the amount of variation within each source. It is a measure of the variability in the data.

3. **Degrees of Freedom (df)**: Degrees of freedom indicate the number of values in the final calculation of a statistic that are free to vary. It helps determine the shape of the distribution used for hypothesis testing.

4. **Mean Square (MS)**: Mean square is calculated by dividing the sum of squares by the degrees of freedom. It represents the variance within each source of variation.

5. **F-statistic**: The F-statistic is calculated by taking the ratio of the mean square for a specific source to the mean square for error (residuals). It measures whether there are statistically significant differences between groups.

6. **p-value**: The p-value associated with each F-statistic is a measure of the significance of that source of variation. It tells you the probability of observing the observed variation (or more extreme) if there were no real effect. Smaller p-values indicate more significant effects.

7. **Significance**: You'll often see a column labeled "PR(>F)" or "Prob (F-statistic)." This column provides the p-value associated with each source of variation.

**Interpretation**:

- **Main Effects**:
  - If the p-value for a main effect (e.g., "Pclass" or "Sex") is small (typically below a chosen significance level, e.g., 0.05), you can conclude that there is a statistically significant effect of that variable on the dependent variable ("Fare"). In other words, the variable has an impact on the "Fare" that is unlikely to be due to random chance alone.

- **Interaction Effect**:
  - If the p-value for the interaction effect ("Pclass:Sex") is small, it indicates that there is a significant interaction between "Pclass" and "Sex." This means that the combined effect of both variables on "Fare" is different from what would be expected based on the individual effects of each variable. In other words, there is a synergistic or interactive effect.

- **Sum of Squares and Mean Square**:
  - The larger the sum of squares for a source of variation relative to the mean square error, the more significant that source is. A large F-statistic with a small p-value suggests a strong effect.

In summary, the ANOVA table allows you to determine whether there are statistically significant differences in the "Fare" based on "Pclass," "Sex," and their interaction. You interpret the results by looking at the p-values associated with each source of variation and comparing them to your chosen significance level. If the p-value is below your significance level, you conclude that the corresponding effect is statistically significant.
