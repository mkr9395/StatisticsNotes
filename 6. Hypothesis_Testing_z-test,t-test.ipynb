## Hypothesis Testing:


Hypothesis testing is a statistical inferential method used to test claims/assumption about the parameters of population to be correct or not with the sample data in hand.



The purpose of hypothesis testing is to gather evidence (data) to either
reject or fail to reject the null hypothesis in favour of the alternative hypothesis, which
claims there is a significant effect or relationship.

### Null and Alternate Hypothesis:

#### Null hypothesis (H0):
In simple terms, the null hypothesis is a statement that assumes there is no significant
effect or relationship between the variables being studied. 


It serves as the starting point
for hypothesis testing and represents the ___status quo___ or the assumption of no effect until
proven otherwise. 



#### Alternative hypothesis (H1 or Ha):
The alternative hypothesis, is a statement that contradicts the null hypothesis and claims
there is a significant effect or relationship between the variables being studied. 


It
represents the ___research hypothesis___ or the claim that the researcher wants to support
through statistical analysis.

### Important Points
How to decide what will be Null hypothesis and what will be Alternate Hypothesis

>__Typically the Null hypothesis says nothing new is happening__


> __We try to gather evidence to reject the null hypothesis__


>__It's important to note that failing to reject the null hypothesis doesn't necessarily mean
that the null hypothesis is true; it just means that there isn't enough evidence to support
the alternative hypothesis.__

Hypothesis tests are similar to jury trials, in a sense. In a jury trial, H0 is similar to the not-guilty verdict,
and Ha is the guilty verdict. You assume in a jury trial that the defendant isnâ€™t guilty unless the
prosecution can show beyond a reasonable doubt that he or she is guilty. If the jury says the evidence is
beyond a reasonable doubt, they reject H0, not guilty, in favour of Ha , guilty.

#### NOTE:  We will either reject the null hypothesis or fail to reject the null hypothesis. We never accept the null hypothesis

### How to decide what will be the null hypothesis:


- The null hypothesis represents the default or status quo assumption, while the alternative hypothesis represents the claim or effect you want to test. 


- The choice of which statement becomes the null hypothesis depends on the research question and the hypothesis you want to investigate. Here's how to decide which statement should be your null hypothesis, illustrated with examples:



**1. The Null Hypothesis is the Status Quo or No Effect Hypothesis:**

   - The null hypothesis typically represents the absence of an effect, no difference, or no relationship. It is often a statement of no change or no effect.
   - The alternative hypothesis, on the other hand, represents the opposite of the null hypothesis and is used to test if there is a significant effect, difference, or relationship.

##### Example: You are testing a new drug's effectiveness in reducing blood pressure.
   - Null Hypothesis (H0): The drug has no significant effect on reducing blood pressure.
   - Alternative Hypothesis (Ha): The drug has a significant effect on reducing blood pressure.


<br></br>

**2. The Null Hypothesis is the Hypothesis of Equality:**

   - In some cases, the null hypothesis may be framed as a statement of equality or no change, and the alternative hypothesis represents inequality or change.
   - This is often used when comparing two groups or populations.

##### Example : You want to compare the mean scores of two groups of students (Group A and Group B) in a mathematics exam.
   - Null Hypothesis (H0): The mean score of Group A is equal to the mean score of Group B.
   - Alternative Hypothesis (Ha): The mean score of Group A is not equal to the mean score of Group B.

<br></br>

**3. The Null Hypothesis Challenges a Claim or Established Theory:**

   - Sometimes, the null hypothesis is used to challenge a previously made claim or established theory. In this case, the null hypothesis represents the rejection of that claim.
   - The alternative hypothesis supports the claim or theory.

##### Example : A car manufacturer claims that their new hybrid car model achieves an average fuel efficiency of 50 miles per gallon (mpg).
   - Null Hypothesis (H0): The average fuel efficiency of the new hybrid car model is 50 mpg.
   - Alternative Hypothesis (Ha): The average fuel efficiency of the new hybrid car model is not 50 mpg.


<br></br>
**4. The Null Hypothesis Represents a Specific Value or Parameter:**

   - In some cases, the null hypothesis may represent a specific value or parameter, and the alternative hypothesis tests whether the parameter differs from that value.
   
##### Example : You are testing whether the population mean IQ score is different from 100.
   - Null Hypothesis (H0): The population mean IQ score is equal to 100.
   - Alternative Hypothesis (Ha): The population mean IQ score is not equal to 100.

### NOTE : the values we obtain from z-test is Z-statistics

A Z-test is a statistical test used to determine whether the means of two data sets are significantly different from each other when you have a large enough sample size and know the population standard deviation. There are a few different variations of the Z-test, each suited for specific situations:



1. **One-Sample Z-Test:** This test is used when you have one data set and want to compare its sample mean to a known population mean. It's typically used when you have a large sample size (typically n > 30) and know the population standard deviation. The formula for the test statistic is:

   
   $$Z-statistics=\frac{(\bar x - \mu)}{\frac{\sigma}{\sqrt n}}$$

   Where:
   - Z is the test statistic.
   - xÌ„ is the sample mean.
   - Î¼ is the population mean.
   - Ïƒ is the population standard deviation.
   - n is the sample size.



2. **Two-Sample (Indepndent) Z-Test:** This test is used when you have two data sets and want to compare their means to determine if they are significantly different. Like the one-sample Z-test, it assumes you know the population standard deviations of both samples. The formula for the test statistic is slightly different:

  
   $$Z-statistics=\frac{(\bar x_1 - \bar x_2 )}{\sqrt{\frac{(\sigma_1^2)}{n_1}+\frac{(\sigma_2^2)}{n_2}}}$$

   Where:
   - Z is the test statistic.
   - xÌ„1 and xÌ„2 are the sample means of the two samples.
   - Ïƒ1 and Ïƒ2 are the population standard deviations of the two samples.
   - n1 and n2 are the sample sizes of the two samples.



3. **Paired (Dependent) Z-Test:** This test is used when you have paired or matched data points. It compares the means of the paired differences to determine if they are significantly different from zero. It's often used in "before and after" or "treatment and control" scenarios. The formula for the test statistic is similar to the one-sample Z-test:

   $$Z-statistics = \frac{(\bar x_d - \mu_d)} {(\sigma_d/\sqrt n)}$$

   Where:
   - Z is the test statistic.
   - xÌ„d is the mean of the paired differences.
   - Î¼d is the hypothesized population mean of the paired differences.
   - Ïƒd is the population standard deviation of the paired differences.
   - n is the number of paired data points.

## Statistical errors - Type 1 vs Type 2 Error :

There are two types of errors we can make while testing:

### Type-I (False Positive) error

__$$\text{trick - (positive means null hypo true, False means you did opposite and rejected it)}$$__


- occurs when the sample results,
lead to the __rejection of the null hypothesis when it is in fact
true.__


- In other words, it's the mistake of finding a significant effect or
relationship when there is none. 


- The probability of committing
a Type I error is denoted by Î± (alpha), which is also known as
the significance level. By choosing a significance level,
researchers can control the risk of making a Type I error.


### Type-II (False Negative) error  $(\beta)$

__$$\text{trick - (Negative means null hypo not true,  False means you did opposite and did not rejected it)}$$__


- occurs when based on the
sample results, the __null hypothesis is not rejected when it is in
fact false.__


- This means that the researcher fails to detect a significant
effect or relationship when one actually exists. The probability
of committing a Type II error is denoted by Î² (beta). $\beta$ depends on sample size $(n)$ and variance $(\sigma)$


- So if your topic is difficult to test due to hard sampling or the effect you are looking for is almost negligible, it is more likely to make this type of error.
    
    
#### **Power of the test:** probability that the test will correctly reject a null hypothesis that is False . $(1-\beta)$
 - Power of test is increased by increasing the sample size.

    Q. You are taking a pregnancy test. The null hypothesis of this test is: I am not pregnant. In reality, you are not pregnant, but the test says you are. Which type of error occurred?

>Ans. This is a Type I error. You rejected a null hypothesis that is true. It is also called a false positive, although in this case, it is counter intuitive. While usually no error is made (thatâ€™s why hypothesis testing is so wide-spread), it is impossible to make both errors at the same time.

### Increasing $\alpha$ rejection region increases and type-1 error increases

### Reducing $\alpha$ rejection region decreases and type-2 error increases

### Trade-off between Type 1 and Type 2 errors:

- The tradeoff between Type I and Type II errors is that reducing the probability of one type of error will increase the probability of the other type of error. 


- In other words, __if we set a stricter significance level (i.e., lower alpha) to reduce the likelihood of Type I errors, we may end up increasing the likelihood of Type II errors. Similarly, if we increase the sample size to reduce the likelihood of Type II errors, we may end up increasing the likelihood of Type I errors.__


- Thus, it is important to strike a balance between the two types of errors based on the context and consequences of the decision being made.

### One tail vs two tail test

#### One-sided (one-tailed) test: 
    
    
- A one-sided test is used when the researcher is interested in
    testing the effect in a specific direction (either greater than or less than the value specified in
    the null hypothesis). The alternative hypothesis in a one-sided test contains an inequality
    (either ">" or "<").


- Example: A researcher wants to test whether a new medication increases the average recovery rate compared to the existing medication.

##### one tailed can be of 2 types:



___1. One-sided right-tailed test :___ It tests if a sample statistic is significantly greater than a hypothesized population parameter; e.g., testing if a student's score is __higher than the passing threshold.__

   - ___any value > critical-value will be in rejection region___

   - ___any value< critical-value will be in rejection region___




#### Two-sided (two-tailed) test: 


- A two-sided test is used when the researcher is interested in
    testing the effect in both directions (i.e., whether the value specified in the null hypothesis is

- Example: A researcher wants to test whether a new medication has a different average

__The main difference between them lies in the directionality of the alternative hypothesis and
how the significance level is distributed in the critical regions.__

### Advantages and Disadvantages

### One-tailed test (one-sided):

__Advantages:__

- __More powerful:__ One-tailed tests are generally more powerful than two-tailed tests, as the
entire significance level (Î±) is allocated to one tail of the distribution. This means that the
test is more likely to detect an effect in the specified direction, assuming the effect exists.


- __Directional hypothesis:__ One-tailed tests are appropriate when there is a strong theoretical
or practical reason to test for an effect in a specific direction. Not equal but don't know it is less or more.


__Disadvantages:__

- __Missed effects:__ One-tailed tests can miss effects in the opposite direction of the specified
alternative hypothesis. If an effect exists in the opposite direction, the test will not be
able to detect it, which could lead to incorrect conclusions.


- __Increased risk of Type I error:__ One-tailed tests can be more __prone to Type I errors__ if the
effect is actually in the opposite direction than the one specified in the alternative
hypothesis.

### Two-tailed test (two-sided):

__Advantages:__

- __Detects effects in both directions__


- __More conservative:__ Two-tailed tests are more conservative because the significance level
(Î±) is split between both tails of the distribution. This reduces the risk of Type I errors in
cases where the direction of the effect is uncertain.


__Disadvantages:__

- __Less powerful:__ Two-tailed tests are generally less powerful than one-tailed tests because
the significance level (Î±) is divided between both tails of the distribution. This means the
test requires a larger effect size to reject the null hypothesis, which could lead to a __higher
risk of Type II error (failing to reject the null hypothesis when it is false).__


- __Not appropriate for directional hypotheses:__ Two-tailed tests are not ideal for cases where
the research question or hypothesis is directional, as they test for differences in both
directions, which may not be of interest or relevance.

### Where can be Hypothesis Testing Applied?

1. __Testing the effectiveness of interventions or treatments:__ Hypothesis testing can be used to
determine whether a new drug, therapy, or educational intervention has a significant effect
compared to a control group or an existing treatment.


2. __Comparing means or proportions:__ Hypothesis testing can be used to compare means or
proportions between two or more groups to determine if there's a significant difference.
This can be applied to compare average customer satisfaction scores, conversion rates, or
employee performance across different groups.


3. __Analysing relationships between variables:__ Hypothesis testing can be used to evaluate the
association between variables, such as the correlation between age and income or the
relationship between advertising spend and sales.


4. Evaluating the goodness of fit: Hypothesis testing can help assess if a particular theoretical
distribution (e.g., normal, binomial, or Poisson) is a good fit for the observed data.


5. Testing the independence of categorical variables: Hypothesis testing can be used to
determine if two categorical variables are independent or if there's a significant association
between them. For example, it can be used to test if there's a relationship between the
type of product and the likelihood of it being returned by a customer.


6. __A/B testing:__ In marketing, product development, and website design, hypothesis testing is
often used to compare the performance of two different versions (A and B) to determine
which one is more effective in terms of conversion rates, user engagement, or other
metrics.

### Hypothesis Testing ML Applications

1. __Model comparison:__ Hypothesis testing can be used to compare the performance of

4. __Assessing model assumptions:__ In some cases, machine learning models rely on certain

Hypothesis testing can help assess whether these assumptions are met, allowing you to
determine if the model is appropriate for the data.

- P-values are used in hypothesis testing to help decide whether to reject the

#### EXAMPLE 1: One-tail z-test:

Suppose a company is evaluating the impact of a new training program on the productivity of its employees. The
company has data on the average productivity of its employees before implementing the training program. The
average productivity was 50 units per day. After implementing the training program, the company measures the
productivity of a random sample of 30 employees. The sample has an average productivity of 53 units per day and
the pop std is 4. The company wants to know if the new training program has significantly increased productivity.

$\mu$ = 50 (population mean)


$\sigma=4$ (population std dev)


n = 30 (sample space)

$\bar x = 53$ (sample mean)

__Step 1 : Formulate a Null and Alternate hypothesis__

- $H_0:$ average productivity is still 50 units per day. $\mu=50$.


- $H_a:$ average productivity increaesed to 53 units per day. so $\mu>50$


__Step 2 : Select a significance level__

- $\alpha = 0.05\;or\;5$%

__Step 3: Check assumptions__

- Normality valid as sample size >=30 and population std dev is known.


__Step 4 : Decide which test is appropriate(Z-test, T-test, Chi-square test, ANOVA)__
- we will use ___one tail z-test___ as population std dev is known.

__Step 5 : State the relevant test statistic__
- as we are performing z-test we will calculate z

__Step 6 : Conduct the test__


- Sample follows normal distribution:

$$z-statistics\;=\;\frac{\bar x - \mu}{\frac{\sigma}{\sqrt n}}$$

$$=\frac{53-50}{\frac{4}{\sqrt 30}} \Longrightarrow 4.10$$

___NOTE : we don't calculate critical value here as before one___

### now the p-value approach

__Step 7 : we will calculate area after the z-statistics in the z-table (right-area for positive values)__

- We want to break the null hypothesis even more extreme.

### NOTE : z-table give area to the left but we need area to the right. so we will do p-value = (1- the value)

$$0.9999$$

$$\text{so p-value = 1-0.9999 = 0.0001}$$

#### the same using python

```python
import numpy as np
from scipy import stats
```

```python
alpha = 0.05
```

```python
population_mean = 50  # Average productivity before training program
sample_mean = 53  # Average productivity after training program
sample_size = 30  # Number of employees in the sample
population_std = 4  # Population standard deviation
```

$$z-statistics\;=\;\frac{\bar x - \mu}{\frac{\sigma}{\sqrt n}}$$

```python
z_stats = (sample_mean - population_mean)/(population_std/np.sqrt(sample_size))
z_stats
```

```python
# Calculate the p-value for a one-tailed test (right-tailed test)
p_value = 1 - stats.norm.cdf(z_stats)
p_value
```

```python
# Print the results
print(f"Z-score: {z_stats}")
print(f"p-value: {p_value:.6f}")
```

```python
# Check if the null hypothesis is rejected
if p_value < alpha:
    print("Reject the null hypothesis. The average productivity has increased to 53 units per day.")
else:
    print("Fail to reject the null hypothesis. There is no significant increase in average productivity.")
```

__Step 8 : Compare p-value with significance level__

- 0.00001 < 0.05. __So we reject the null hypothesis. so the average productivity has increased from 50 units__

A t-test is a statistical test used in hypothesis testing to compare the means of two samples or to compare a sample mean to a known population mean. 


__The t-test is based on the t-
distribution, which is used when the population standard deviation is unknown and the sample size is small.__

### There are three main types of t-tests:

### 1. One-sample t-test: 

- The one-sample t-test is used to compare the mean of a single sample to a
known population mean. 


- The null hypothesis states that there is no significant difference
between the sample mean and the population mean, while the alternative hypothesis states
that there is a significant difference.
   Formula:
   $$
   t = \frac{{\bar{X} - \mu}}{{s / \sqrt{n}}}
   $$

   - t is the t-statistic.
   - $\bar{X}$ is the sample mean.
   - $\mu$ is the population mean or hypothesized mean.
   - s is the sample standard deviation.
   - n is the sample size.

### 2. Independent two-sample t-test (var equal): 

- The independent two-sample t-test is used to compare the
means of two independent samples. 


- The null hypothesis states that there is no significant difference between the means of the two samples, while the alternative hypothesis states that there is a significant difference. eg: same student can't be in 2 sections.
$$t \Longrightarrow \frac{\bar x_1 - \bar x_2}{\sqrt{\frac{{s_1}^2}{n_1}+ \frac{{s_2}^2}{n_2}}} $$

   - t is the t-statistic.
   - $\bar{x}_1$ and $\bar{x}_2$ are the sample means of the two groups.
   - $s_1$ and $s_2$ are the sample standard deviations of the two groups.
   - $n_1$ and $n_2$ are the sample sizes of the two groups.
##### NOTE : Degree of freedom for 2 sample t-test is (n1 + n2 - 2)

### 3. Paired t-test (dependent two-sample t-test): 

- The paired t-test is used to compare the means of
two samples that are dependent or paired, such as __pre-test and post-test scores for the same
group of subjects or measurements taken on the same subjects under two different
conditions.__ 


- The null hypothesis states that there is no significant difference between the
means of the paired differences, while the alternative hypothesis states that there is a
significant difference.

Formula:
   $$
   t = \frac{{\bar{X}_d - \mu_d}}{\frac{s_d}  {\sqrt{n}}}
   $$

   - t is the t-statistic.
   - $\bar{X}_d$ is the mean of the differences between paired observations.
   - $\mu_d$ is the population mean of the differences (usually 0 under the null hypothesis).
   - $s_d$ is the standard deviation of the differences.
   - n is the number of paired observations.

An independent two-sample t-test, also known as an unpaired t-test, is a statistical method
used to compare the means of two independent groups to determine if there is a significant
difference between them.

#### Assumptions for the test:

1. __Independence of observations:__ The two samples must be independent, meaning there is
no relationship between the observations in one group and the observations in the other
group. The subjects in the two groups should be selected randomly and independently.


2. __Normality:__ The data in each of the two groups should be approximately normally
distributed. The t-test is considered robust to mild violations of normality, especially
when the sample sizes are large (typically n â‰¥ 30) and the sample sizes of the two groups are similar. If the data is highly skewed or has substantial outliers, consider using a non-
parametric test, such as the Mann-Whitney U test.



3. __Equal variances (Homoscedasticity):__ The variances of the two populations should be
approximately equal. This assumption can be checked using F-test for equality of
variances. If this assumption is not met, you can use Welch's t-test, which does not
require equal variances.


4. __Random sampling:__ The data should be collected using a random sampling method from
the respective populations. This ensures that the sample is representative of the
population and reduces the risk of selection bias.


### case1 : T-statistics when population variances or standard deviations are equal

In case, __the two populationsâ€™ variances or standard deviations are equal__, the formula termed as pooled t-statistics is used based on the usage of pooled standard deviations of the two samples. The following is the formula for the pooled t-statistics:

### But we calculate pooled std dev $s_p$

- __degree of freedom = n1 + n2 - 2__

### standard error formula : 

 Although using pooled variance in a tâ€test is generally more likely to yield significant results than using separate variances, it is often hard to know whether the variances of the two populations are equal. For this reason, the pooled variance method should be used with caution. 

### When to use pooled std dev or not?


In a two-sample independent t-test, whether you should use the pooled standard deviation (pooled standard error) or not depends on whether you assume equal variances (homoscedasticity) between the two groups. Here's when to use it and when not to use it:

**Use Pooled Standard Deviation (Equal Variances Assumed):**

1. **Equal Variances Assumption:** Use the pooled standard deviation when you assume that the population variances of the two groups you're comparing are approximately equal. In other words, you believe that the variances of the two populations are not significantly different.

2. **Equal Sample Sizes:** It is typically used when you have roughly equal sample sizes in both groups. This is not a strict requirement, but it's more common to use the pooled standard deviation when sample sizes are similar.

3. **Balanced Design:** A balanced design means that you have roughly the same number of observations in each group.

**Do Not Use Pooled Standard Deviation (Equal Variances Not Assumed):**

1. **Unequal Variances Assumption:** If you have reason to believe that the population variances of the two groups are significantly different, you should not use the pooled standard deviation. This is called the "heteroscedasticity" case.

2. **Unequal Sample Sizes:** Even if you have roughly equal sample sizes, if you suspect that the population variances are different, you should not use the pooled standard deviation.

3. **Unbalanced Design:** An unbalanced design means you have significantly different sample sizes in the two groups. In this case, you should be cautious about assuming equal variances unless you have strong evidence to support it.

In practice, it's a good idea to conduct a preliminary test for homogeneity of variances, such as Levene's test or Bartlett's test, to assess whether the assumption of equal variances holds. If the test suggests that the variances are significantly different, you should use a modification of the t-test that does not assume equal variances, such as Welch's t-test.

So, to summarize, use the pooled standard deviation when you reasonably assume equal variances and have a relatively balanced design. Otherwise, consider using an alternative approach that does not assume equal variances, such as Welch's t-test.

### case2 : Welch's t-test $\Longrightarrow$ T-statistics when homogeneity of variances does not hold good (variances not equal):

When the __variances of populations are not equal__, the following formula is used to calculate the T-statistics and degrees of freedom.

### Here as variances not equal, we will apply welch test to find degree of freedom:

 The formula used to calculate \(df\) for Welch's t-test is:

$$ df = \frac{\left(\frac{{s_1^2}}{{n_1}} + \frac{{s_2^2}}{{n_2}}\right)^2}{\frac{{\left(\frac{{s_1^2}}{{n_1}}\right)^2}}{{n_1 - 1}} + \frac{{\left(\frac{{s_2^2}}{{n_2}}\right)^2}}{{n_2 - 1}}} $$

Where:

- $s_1^2$ and $s_2^2$ are the sample variances of the two independent samples.
- $n_1$ and $n_2$ are the sample sizes of the two independent samples.

## All about t.ppf , t.pmf, t.cdf :

### Q - Difference between `t.ppf` and `t.cdf`:

In the context of a t-test and hypothesis testing, `t.ppf` and `t.cdf` are both functions used to work with the t-distribution, but they serve different purposes:

In simple terms:

1. **Percent Point Function (PPF):**
   - Use PPF when you have a specific probability (a percentile or quantile) and you want to find the corresponding value from a probability distribution.
   - It answers questions like, "What value marks the 95th percentile of this distribution?" or "What's the critical value that separates the lowest 10% of the distribution?"
   - PPF takes a probability as input and gives you a value as output.



2. **Cumulative Distribution Function (CDF):**
   - Use CDF when you have a specific value and you want to find the probability of getting a value less than or equal to that specific value in a probability distribution.
   - It answers questions like, "What is the probability that a random observation from this distribution is less than or equal to a given value?"
   - CDF takes a value as input and gives you a probability as output.

In summary, PPF is used to find values corresponding to specific probabilities, while CDF is used to find probabilities corresponding to specific values. These functions are commonly used in statistics to make inferences and perform calculations with probability distributions.





#### Interpretation Related to p-value and Critical Value:

- **Critical Value (using `t.ppf`):** When performing a hypothesis test, you calculate a test statistic (t-statistic) from your data. You then use `t.ppf` to find the critical value(s) based on your chosen significance level (alpha) and degrees of freedom. If your t-statistic falls beyond the critical value(s), you reject the null hypothesis. The critical value(s) help define the boundaries for the rejection region.



- **p-value (using `t.cdf`):** To calculate the p-value associated with your t-statistic, you use `t.cdf`. The p-value represents the probability of observing a t-statistic as extreme as the one you obtained (or more extreme) under the null hypothesis. If the p-value is less than your chosen significance level (alpha), you reject the null hypothesis. A smaller p-value indicates stronger evidence against the null hypothesis.
<br></br>
In summary, `t.ppf` helps you find critical values that define the rejection region, while `t.cdf` helps you calculate the p-value associated with your test statistic. Both are crucial in hypothesis testing, where you compare the observed test statistic to these values to make a decision about the null hypothesis.

### Q - Difference between `t.ppf` and `t.pmf`:

In simple terms:

1. **Percent Point Function (PPF):**
   - Use PPF when you have a specific probability (a percentile or quantile) and you want to find the corresponding value from a probability distribution.
   - It answers questions like, "What value marks the 95th percentile of this distribution?" or "What's the critical value that separates the lowest 10% of the distribution?"
   - PPF takes a probability as input and gives you a value as output.


2. **Probability Mass Function (PMF):**
   - Use PMF when you want to find the probability of a specific discrete outcome (a specific value) occurring in a probability distribution.
   - It answers questions like, "What is the probability of getting exactly 3 heads when tossing a fair coin 5 times?"
   - PMF takes a value as input (one of the possible discrete outcomes) and gives you the probability of that specific outcome occurring.

In summary, PPF is used to find values corresponding to specific probabilities, while PMF is used to find the probability of specific discrete outcomes in a probability distribution, typically for discrete random variables like coin tosses or dice rolls.

### Q- Calculation of `t.cdf` in t-test for one-tailed left (less than) and right(greater than):

To calculate the p-value for the "greater than" and "less than" directions in a one-tailed t-test using `t.cdf`, you can use the following formulas:

**For the "Greater Than" Direction:**
- If you want to test whether a sample mean is significantly greater than a hypothesized population mean $(H_0: \mu \leq \text{hypothesized value})$, you calculate the p-value for the "greater than" direction.

The p-value formula for the "greater than" direction using `t.cdf` is:
$$p_{\text{greater}} = 1 - \text{t.cdf}(t_{\text{observed}}, \text{df})$$
Where:
- $p_{\text{greater}}$ is the p-value for the "greater than" direction.

$$p_{\text{less}} = \text{t.cdf}(t_{\text{observed}}, \text{df})$$
Where:
- $p_{\text{less}}$ is the p-value for the "less than" direction.
- $t_{\text{observed}}$ is the observed t-statistic.
- $\text{df}$ is the degrees of freedom.
- $\text{t.cdf}(t_{\text{observed}}, \text{df})$ calculates the cumulative probability up to $t_{\text{observed}}$ in the t-distribution.

In both cases, you can compare the calculated p-value to your chosen significance level (alpha) to make a decision about rejecting or failing to reject the null hypothesis. If the p-value is less than or equal to alpha, you reject the null hypothesis; otherwise, you fail to reject it.

### Q - how do we calculate z-test : 

for z-test we use `stats.norm.ppf(alpha)`

```python
import scipy.stats as stats

# Sample data
sample_mean = 168  # Sample mean height
population_std = 8  # Population standard deviation
sample_size = 50  # Sample size

# Null hypothesis: The average height is 165 cm (H0: Î¼ = 165)
# Alternative hypothesis: The average height is not 165 cm (Ha: Î¼ â‰  165)

# Significance level (alpha)
alpha = 0.05

# Calculate the test statistic (z-score)
z = (sample_mean - 165) / (population_std / (sample_size**0.5))

# Calculate the critical values (two-tailed test)
critical_value_left = stats.norm.ppf(alpha / 2)
critical_value_right = stats.norm.ppf(1 - alpha / 2)

# Make a decision
if z < critical_value_left or z > critical_value_right:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

# Draw a conclusion
if z < critical_value_left:
    print("The average height is significantly less than 165 cm.")
elif z > critical_value_right:
    print("The average height is significantly greater than 165 cm.")
else:
    print("There is no significant difference in average height from 165 cm.")
```
