- Normal distribution / Gaussian distribution, is a continuous probability distribution that is symmetrical around the mean, with a bell-shaped curve.


- In simple terms, it means that most data points cluster around the mean (average) value, with fewer data points falling farther away from the mean in a predictable pattern.

$$Denoted\;by\;\; N(\mu,\sigma^2)$$
$$X \sim N(\mu,\sigma^2)$$
variable X follows a normal distribution with mean mu and variance sigma squared.

Commonality in Nature: Many natural phenomena follow a normal distribution, such as
the heights of people, the weights of objects, the IQ scores of a population, and many
more. Thus, the normal distribution provides a convenient way to model and analyse such
data.

### Characteristics:
* The graph of normal distribution is bell shaped, so majority of the data is centred around the mean. thus values further away from the mean are less likely to occur.
* we can see that the graph is symmetric with regards to the mean. That suggests values equally far away in opposing directions would still be equally likely.

$$Expected\;value\;\;E(x) = \mu$$

$$Var(X) = \sigma^2\;\;or\;\;Var(X) = E(X^2)-E(X)^2$$

>**The 68, 95, 99.7 Law**
<br></br>
This law suggests that for any normally distributed event, 68 percent of all outcomes fall within one standard deviation away from the mean. 95 percent fall within two standard deviations and 99.7 percent within three.


### Transformation
A transformation is a way in which we can alter every element of a distribution to get a new distribution with similar characteristics. For normal distributions, we can use addition, subtraction, multiplication and division without changing the type of the distribution.

* Addition : Adding a number to the entire distribution shifts the graph  to the right
* Substraction : Subtracting a number from every element of the  distribution shifts the graph to the left
* multiplication : Multiplying the function by a constant, it will shrink the graph that many times 
* Division : Dividing  every element Expands the graph

>if we multiply or divide by a number between 0 and 1 , the opposing effects will occur. For example, dividing by a half is the same as multiplying by two, so the graph will shrink even though we are dividing.

#### - Mean increase then graph moves towards right.


#### - Mean decrease then graph moves left.

__A z-table tells you the area under a normal distribution curve,__ to the left of the z-score.

#### Here's how Z-tables are used:

1. **Standardization**:
   - Convert your raw data values from a normal distribution with any mean and standard deviation to the standard normal distribution with a mean of 0 and a standard deviation of 1. This process is called standardization and is done using the formula: 

     **Z = (X - ) / **

   - Here, X represents your original data point,  is the mean of the original distribution, and  is the standard deviation of the original distribution.


2. **Lookup Z-Score**:
   - After standardizing your data, you obtain a Z-score (Z) that represents how many standard deviations your data point is from the mean. This Z-score is what you use to look up values in the Z-table.



3. **Finding Probabilities**:
   - In the Z-table, you'll find the probability associated with the Z-score you calculated. The table provides the cumulative probability (also known as the percentile or area under the curve) for Z-scores. This represents the probability that a random variable from the standard normal distribution is less than or equal to the Z-score.



4. **Interpolation**:
   - Sometimes, the exact Z-score you calculated might not be in the table. In such cases, you can interpolate between the closest values in the table to estimate the probability associated with your Z-score.



5. **Inverse Lookup** (Optional):
   - You can also use Z-tables to find the Z-score corresponding to a given probability. For example, if you want to find the Z-score that corresponds to the 95th percentile, you can use the table in reverse to find the Z-score associated with that probability.

## Skewness

The direction of the skew is counterintuitive. It does not depend on which side the line is leaning to, but rather to which side its tail is leaning to.
<br></br>


* __Left (negative) skewed:__ It means that the outliers are to the left (long tail to the left). $i.e.\;mean<median$


* __Centre skewed:__ Data is normally/equally distributed. $i.e.\; mean = mode = median$



* __Right (positive) skewed:__ It means that the outliers are to the right (long tail to the right).
$i.e.\;mean>median$

## Kurtosis

Kurtosis is the 4th statistical moment. In probability theory and statistics, kurtosis
(meaning "curved, arching") is a measure of the "tailedness" of the probability
distribution of a real-valued random variable. Like skewness, kurtosis describes a
particular aspect of a probability distribution.


__Kurtosis refers to the proportion of data that is heavy-tailed or light-tailed in comparison with a normal distribution. Density of outliers.__

### NOTE : Kurtosis doesnot deals with peakedness. It only deals with fattness of tails (presence of outliers). 

### Formula for Sample Kurtosis:



### Practical Use-case
In finance, ___kurtosis risk___ refers to the risk associated with the possibility of extreme
outcomes or "fat tails" in the distribution of returns of a particular asset or portfolio.
If a distribution has high kurtosis, it means that there is a higher likelihood of
extreme events occurring, either positive or negative, compared to a normal
distribution.


In finance, kurtosis risk is important to consider because it indicates that there is a
greater probability of large losses or gains occurring, which can have significant
implications for investors. As a result, investors may want to adjust their investment
strategies to account for kurtosis risk.

#### Excess Kurtosis & Types
Excess kurtosis is a measure of how much more peaked or flat a distribution is
compared to a normal distribution, which is considered to have a kurtosis of 0. It is
calculated by subtracting 3 from the sample kurtosis coefficient.

### Types of Kurtosis:

### 1. Leptokurtic

A distribution with positive excess kurtosis is called leptokurtic. "Lepto-" means
"slender". In terms of shape, a leptokurtic distribution has fatter tails. __This indicates
that there are more extreme values or outliers in the distribution.__


Example - Assets with positive excess kurtosis are riskier and more volatile than those
with a normal distribution, and they may experience sudden price movements that
can result in significant gains or losses.


### 2. Platykurtic

A distribution with negative excess kurtosis is called platykurtic. "Platy-" means
"broad". In terms of shape, a platykurtic distribution has thinner tails. __This indicates
that there are fewer extreme values or outliers in the distribution.__

Assets with negative excess kurtosis are less risky and less volatile than those with a
normal distribution, and they may experience more gradual price movements that
are less likely to result in large gains or losses.


### 3. Mesokurtic


Distributions with zero excess kurtosis are called mesokurtic. The most prominent
example of a mesokurtic distribution is the normal distribution family, regardless of
the values of its parameters.


Mesokurtic is a term used to describe a distribution with a excess kurtosis of 0,
indicating that it has the same degree of "peakedness" or "flatness" as a normal
distribution.

Example -
In finance, a mesokurtic distribution is considered to be the ideal distribution for
assets or portfolios, as it represents a balance between risk and return.

### <span class="girk">Interview Question : How to find if a given distribution is normal or not?</span>

- __Visual inspection:__ One of the easiest ways to check for normality is to visually inspect
a histogram or a density plot of the data. A normal distribution has a bell-shaped
curve, which means that the majority of the data falls in the middle, and the tails
taper off symmetrically. If the distribution looks approximately bell-shaped, it is likely
to be normal.


- __<span class="girk">QQ Plot:</span>__ Another way to check for normality is to create a normal probability plot
(also known as a Q-Q plot) of the data. A normal probability plot plots the observed
data against the expected values of a normal distribution. If the data points fall along
a straight line, the distribution is likely to be normal.


- __Statistical tests:__<br></br>
    - There are several statistical tests that can be used to test for normality, such as the __Shapiro-Wilk test, the Anderson-Darling test, and the Kolmogorov-Smirnov test.__ These tests compare the observed data to the expected
values of a normal distribution and provide a p-value that indicates whether the data
is likely to be normal or not.<br></br>
    - __A p-value less than the significance level (usually 0.05)
suggests that the data is not normal.__

## QQ Plot

A QQ plot (quantile-quantile plot) is a graphical tool used to assess the similarity of
the distribution of __two sets of data.__ It is particularly useful for determining whether a
set of data follows a normal distribution.

___It compares 2 columns. In qqplot we already know that one column is normally distributed and we compare the other column with it. If the percentile points formed from both columns form a staright line then it is normally distributed, otherwise not___

In a QQ plot, the quantiles of the two sets of data are plotted against each other. The
quantiles of one set of data are plotted on the x-axis, while the quantiles of the other
set of data are plotted on the y-axis. 

If the two sets of data have the same
distribution, the points on the QQ plot will fall on a straight line. 

If the two sets of
data do not have the same distribution, the points will deviate from the straight line.

## <span class="mark">Pareto Distribution:</span>

### What is Power Law

In mathematics, a power law is a functional relationship between two variables, where one
variable is proportional to a power of the other. Specifically, if y and x are two variables
related by a power law, then the relationship can be written as:

$$y = k * x^a$$


__The "80-20 rule" which says that 20% of the population controls 80% of the wealth__

#### The probability density function (PDF) of the Pareto distribution is given by:

$$
f(x; x_m, \alpha) = 
\begin{cases}
\frac{\alpha x_m^\alpha}{x^{\alpha+1}}, & \text{for } x \geq x_m \\
0, & \text{for } x < x_m
\end{cases}
$$

Where:


- (x) is the random variable (usually representing a positive real number).


- $(x_m)$ is the minimum possible value of $(x)$ (the lower bound of the distribution).


- $(\alpha)$ is the shape parameter, also known as the Pareto index or tail index. It is a positive real number.

This formula defines the probability of observing a value $(x)$ in the Pareto distribution given the parameters $(x_m)$ and $(\alpha)$. The distribution is defined for $(x \geq x_m)$ because it is a right-tailed distribution; it has a lower bound at $(x_m)$, and it extends to positive infinity.

The Pareto distribution is often used to model distributions with heavy tails, where most of the probability mass is concentrated at the lower end, and there is a small but non-negligible probability of observing very large values. It is commonly used in economics, finance, and other fields to model phenomena like income distribution, wealth distribution, and the distribution of city sizes.

- It follows Pareto Distribution





__Higher the alpha value, thinner the tail. Lower the alpha value, fatter the tail__

Examples

- The sizes of human settlements (few cities, many hamlets/villages)


- File size distribution of Internet traffic which uses the TCP protocol (many smaller files,
few larger ones)

## Bernoulli Distribution

Bernoulli distribution is a probability distribution that models a __binary outcome,__ where the
outcome can be either success (represented by the value 1) or failure (represented by the
value 0)


Conditions for Bernoulli Distribution:
- 1 Trial
- 2 Possible outcomes
- eg: tossing a coin, marking True or False
<br></br>
We often have to assign which outcome is 0 and which outcome is 1. 

$$0\leq\;p\;\leq1$$

$$q\;=\;1\;-\;p$$



After doing so, we can calculate the expected value i.e $E(X) = p\;or\;E(X) = 1-p$
<br></br>
Conventionally, we also assign a value of 1 to the event with the probability equal to P and 0 to 1-p. So, E(X) expresses the likelihood of the favoured Event. Since we only have one trial and a favorite event. We expect that outcome to occur. i.e $E(X) = p$

$$\text{Variance of bernoulli distribution is always} = p(1-p)$$

__The Bernoulli distribution is characterized by a single parameter, which is the probability of
success, denoted by p.__

The Bernoulli distribution is commonly used in machine learning for modelling
binary outcomes, such as whether a customer will make a purchase or not,
whether an email is spam or not, or whether a patient will have a certain disease
or not.

### As Bernoulli has discrete outcomes so it will have Probability Mass function(PMF)

#### PMF formula for Bernoulli:

$$P(X = x) = 
\begin{cases} 
1-p & \text{if } x = 0 \\
p & \text{if } x = 1 
\end{cases}$$

```
P(X = 0) = 1 - p
P(X = 1) = p
```

#### CDF (Cumulative Distribution Function) for Bernoulli :

$$
F(x) = 
\begin{cases} 
0 & \text{if } x < 0 \\
1 - p & \text{if } 0 \leq x < 1 \\
1 & \text{if } x \geq 1
\end{cases}
$$

```
F(x) = 0   if x < 0
F(x) = 1 - p   if 0 <= x < 1
F(x) = 1   if x >= 1
```

## Binomial Distribution

Binomial distribution is a probability distribution that describes the number of
successes in a fixed number of independent Bernoulli trials with two possible
outcomes (often called "success" and "failure"), where the probability of success
is constant for each trial. 

The binomial distribution is characterized by two
parameters: the number of trials n and the probability of success p.

- Binomial Distribution is a sequence of identical Bernoulli events.


- It is repeated for fixed number of times.


- The trials are independent and have 2 mutually exclusive outcomes either success or failure.


- The probability of success is same for all trials

$$It\;is\;denoted\;by "B"$$ 

$$B\;(n,p)$$ 

$$X \sim B\;(10,0.6)$$

We read the following statement as variable X follows a binomial distribution with 10 trials and a likelihood of 0.6 to succeed on each individual trial i.e it's not biased


### **Probability distribution function :**

>$$p(y) = {n \choose x}*p^x*(1-p)^{n-x}$$
 OR
 
>$$p(y)\;=\;{}^nC_x*p^x*(1-p)^{n-x}$$


Here $n\;\longrightarrow$ is number of trials and $p \longrightarrow$ is success probability for each trials.
<br></br>

#### NOTE : We can express a Bernoulli distribution as a binomial distribution with a single trial.

$$Bern(p)=Bern(1,p)$$

___eg: there is a 60% i.e 0.6 chance of general motor share to increase.___

___So to calculate 3 increase in 5 days:___
    
$y \rightarrow 3 , $ $n \rightarrow 5 , $ $p \rightarrow 0.6$
    
$$p(x) = 5^C_3 * 0.6^3 * 0.4^2 \;\; = \;0.3456\;or\;34.56 \% $$

There is 34.56% chance of increase in stock value 3 out of 5 days.

### Examples of Binomial Distribution : 


**Example 1: Coin Flips**

- __Situation: You flip a fair coin 5 times, and you want to know the probability of getting exactly 3 heads.__


- Parameters: Number of trials (n) = 5, Probability of success (p) = 0.5 (because a fair coin has a 50% chance of landing heads).


- Calculation: You can use the Binomial probability formula:

  $$P(X = x) = \binom{n}{x} \cdot p^x \cdot (1-p)^{n-x}$$

  - $(x)$ is the number of successes you want (3 in this case).
  
  
  - $\binom{n}{x}$ represents the number of ways to choose $(x)$ successes out of $(n)$ trials, which is calculated as $\binom{5}{3} = 10$.
  
  
  - Plug in the values: $(P(X = 3) = 10 \cdot (0.5)^3 \cdot (0.5)^2 = 0.3125)$.

**Example 2: Exam Questions**

- __Situation: In a multiple-choice exam with 10 questions, each with 4 choices, you randomly guess the answers. You want to find the probability of getting exactly 3 correct answers.__


- Parameters: Number of trials (n) = 10 (number of questions), Probability of success (p) = 0.25 (because there is a 1 in 4 chance of guessing correctly).


- Calculation: Again, use the Binomial probability formula:

  $$P(X = x) = \binom{n}{x} \cdot p^x \cdot (1-p)^{n-x}$$

  
  - x is the number of correct answers you want (3 in this case).
  
  
  - Calculate $\binom{10}{3} = 120$ ways to choose 3 correct answers out of 10.
  
  
  - Plug in the values: $P(X = 3) = 120 \cdot (0.25)^3 \cdot (0.75)^7 \approx 0.250$

**Example 3: Defective Products**

- __Situation: In a production line, each product has a 10% chance of being defective. You want to find the probability that in a batch of 20 products, exactly 2 are defective.__


- Parameters: Number of trials (\(n\)) = 20, Probability of success (\(p\)) = 0.10 (because a 10% chance of being defective).


- Calculation: Use the Binomial probability formula as before:

  $$P(X = x) = \binom{n}{x} \cdot p^x \cdot (1-p)^{n-x}$$

  
  - (x) is the number of defective products you want (2 in this case).
  
  
  - Calculate $(\binom{20}{2} = 190)$ ways to choose 2 defective products out of 20.
  
  
  - Plug in the values: $P(X = 2) = 190 \cdot (0.10)^2 \cdot (0.90)^{18} \approx 0.285$.

## **Difference between bernoulli and Binomial Distribution:**
    
    In a test of 10 questions with only True or False.

- Bernoulli $\rightarrow$ Guessing 1 question
    
    - $E(Bernoulli\;event)\rightarrow$ Which outcome we expect for single trial.


- Binomial $\rightarrow$ Guessing all 10 questions
    
    - $E(Binomial\;event)\rightarrow$ The number of times we expect to get a specific outcome.

### Criteria for Binomial Distribution

- 1. The process consists of n trials


- 2. Only 2 exclusive outcomes are possible, a success and a failure.


- 3. P(success) = p and P(failure) = 1-p and it is fixed from trial to trial


- 4. The trials are independent.

#### Applications of  Binomial Distribution:

1. ___Binary classification problems:___ In binary classification problems, we often model
the probability of an event happening as a binomial distribution. For example, in a
spam detection system, we may model the probability of an email being spam or
not spam using a binomial distribution.


2. ___Hypothesis testing:___ In statistical hypothesis testing, we use the binomial
distribution to calculate the probability of observing a certain number of
successes in a given number of trials, assuming a null hypothesis is true. This can
be used to make decisions about whether a certain hypothesis is supported by
the data or not.


3. ___Logistic regression:___ Logistic regression is a popular machine learning algorithm
used for classification problems. It models the probability of an event happening
as a logistic function of the input variables. Since the logistic function can be
viewed as a transformation of a linear combination of inputs, the output of
logistic regression can be thought of as a binomial distribution.


4. ___A/B testing:___ A/B testing is a common technique used to compare two different
versions of a product, web page, or marketing campaign. In A/B testing, we
randomly assign individuals to one of two groups and compare the outcomes of interest between the groups. Since the outcomes are often binary (e.g., click- through rate or conversion rate), the binomial distribution can be used to model the distribution of outcomes and test for differences between the groups.
